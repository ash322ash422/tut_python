## Union of Two Sets
Question

Given two sets A and B, find all elements that belong to either set.

```python
A = {1, 2, 3, 4, 5}
B = { 4, 5, 1, 7}
```

```python
result = A | B
```

## Intersection of Two Sets
 Question

Find elements that are common to both sets.

```python
A = {1, 2, 3}
B = {2, 3, 4}
```

```python
result = A & B
```

## Difference of Sets
Question

Find elements present in A but not in B.

```python
A = {1, 2, 3, 4}
B = {3, 4}
```

```python
result = A - B
```

## Symmetric Difference
Question

Find elements that belong to exactly one of the sets.

```python
A = {1, 2, 3}
B = {3, 4, 5}
```

```python
result = A ^ B
```

## Subset Check
Question

Check whether set A is a subset of set B.

```python
A = {1, 2}
B = {1, 2, 3, 4}
```

```python
result = A <= B
```

## Cardinality Comparison
Question

Check whether set A has more elements than set B.

```python
A = {1, 2, 3, 4}
B = {5, 6}
```

```python
result = len(A) > len(B)
```

## Universal Difference
Question

Given a universal set U and subset A, find elements not in A.

```python
U = {1, 2, 3, 4, 5}
A = {2, 4}
```

```python
result = U - A
```

## Common Elements Across Containers
Problem

Find elements common to:

a list

a tuple

a set

```python
lst = [1, 2, 3, 4]
tpl = (2, 3, 5)
st  = {3, 4, 5}
```

```python
result = set(lst) & set(tpl) & st
result
```

## List Elements Not in Tuple
Problem

Find elements present in the list but not present in the tuple.

```python
lst = [1, 2, 3, 4, 5]
tpl = (2, 4)
```

```python
result = list(set(lst) - set(tpl))
```

## Frequency Check Using Set Logic
Problem

Check whether all elements in a tuple are unique.

```python
tpl = (1, 2, 3, 4, 5)
```

```python
result = len(tpl) == len(set(tpl))
```

## Mathematical Set Equality
Problem

Check whether a list and tuple contain the same elements, regardless of order.

```python
lst = [1, 2, 3]
tpl = (3, 2, 1)
```

```python
result = set(lst) == set(tpl)
```

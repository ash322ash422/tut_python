# List Comprehensions

## Basic List Comprehension

```python
# Traditional way
squares = []
for i in range(10):
    squares.append(i ** 2)

print("Using loop:", squares)
```

```python
# List comprehension way
squares_comp = [i ** 2 for i in range(10)]

print("Using list comprehension:", squares_comp)
```

## With Condition (Filtering)

```python
# Get even numbers
evens = [x for x in range(20) if x % 2 == 0]

print("Even numbers:", evens)
```

```python
# Extract vowels from string
text = "Artificial Intelligence"
vowels = [c for c in text if c.lower() in "aeiou"]

print(vowels)
```

## Conditional Expression (if-else)

```python
# Label numbers as Even or Odd
labels = ["Even" if x % 2 == 0 else "Odd" for x in range(10)]

print(labels)
```

# Nested List Comprehensions

## Creating a Matrix

```python
# 3x3 matrix
matrix = [[i + j for j in range(3)] for i in range(3)]

print(matrix)
```

#  Shallow Copy vs Deep Copy

- VERY important concept

```python
## Example Data

import copy

original = [[1, 2], [3, 4]]

## Shallow Copy
shallow = copy.copy(original)

# Modify inner list
shallow[0][0] = 999

print("Original:", original)
print("Shallow :", shallow)
```

### Observation:

Both change!

Why?
Because inner objects are shared.


```python
## Deep Copy

original = [[1, 2], [3, 4]]

deep = copy.deepcopy(original)

deep[0][0] = 999

print("Original:", original)
print("Deep Copy:", deep)
```

### Observation

Only deep copy changes.

# Performance Considerations
- **Loop vs List Comprehension Speed**

```python
import time

# Using loop
start = time.time()

squares = []
for i in range(30_000_000):
    squares.append(i*i)

end = time.time()

print("Loop time:", end - start)
```

```python
# Using list comprehension
start = time.time()

squares = [i*i for i in range(30_000_000)]

end = time.time()

print("List comprehension time:", end - start)
```

### Usually list comprehension is **faster**.

```python
## Memory Efficient Alternative — Generator

# Generator expression (uses less memory)
gen = (i*i for i in range(1_000_000))

print(next(gen))
print(next(gen))
```

## Assignment Operators ( =, +=, −=, *=, /=, //=, %=, **= )

### 1 Increment Without Using `+`

Start with `x = 7`.
Update `x` to become `x³` using **only assignment operators**.

---


```python
x = 7
x **= 3
x
```

### 2 Digit Removal

Given `n = 5678`, remove the **last digit** and store the result back in `n`.

---


```python
n = 5678
n //= 10
```

### 3 Swap Without Temp Variable

Given `a = 5` and `b = 9`, swap their values using **only assignment operators**.

---

```python
a = 5
b = 9

a += b
b = a - b
a -= b
```

### 4 Compound Growth

Given `amount = 1000`, increase it by:

* 10%
* then square the result
  (using assignment operators only)

---


```python
amount = 1000
amount *= 1.10
amount **= 2
```

### 5 Reduce to Last Digit

Given any number `n`, update it so that it contains **only its last digit**.

---


```python
n = 7463
n %= 10
```

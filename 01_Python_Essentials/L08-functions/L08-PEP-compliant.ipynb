# PEP compliant code

PEP stands for Python Enhancement Proposal.

It is a design document that describes new features, improvements, or guidelines for the Python programming language.

```python
# 1) Non PEP compliant
def power(base, exponent=2):
    ans = base ** exponent
    return ans

res = power(3, 4)
print(f"The 3 to the power of 4 is {res}")
```

```python
# 2) PEP compliant
def power(base: float, exponent: float = 2) -> float:
    """
    Compute the power of a number.

    This function raises the given base to the specified exponent.
    If no exponent is provided, the function returns the square of the base.

    Parameters
    ----------
    base : float
        The base number.

    exponent : float, optional
        The exponent value. Default is 2.

    Returns
    -------
    float
        The result of base raised to the power of exponent.
    """
    return base ** exponent
```

```python
# 2)
def rectangle_properties(length: float, width: float) -> tuple[float, float, float]:
    """
    Calculate the area, perimeter, and price of a rectangle.

    The price is determined based on the area:
    - If the area is less than 100, price = 2000 × area
    - Otherwise, price = 3000 × area

    Parameters
    ----------
    length : float
        The length of the rectangle.

    width : float
        The width of the rectangle.

    Returns
    -------
    tuple[float, float, float]
        A tuple containing:
        - area of the rectangle
        - perimeter of the rectangle
        - calculated price
    """
    area = length * width
    perimeter = 2 * (length + width)

    if area < 100:
        price = 2000 * area
    else:
        price = 3000 * area

    return area, perimeter, price


# Call the function and print the result
result = rectangle_properties(5, 10)
print(result)
```

```python
# 1) Without PEP: Variable length argument functions

# function defination
def print_everything(*args):
    for a in args:
        print("You entered:", a)

# function calling
print_everything("Python", "Java", "C++", "Rust")
```

```python
# 2) PEP compliant

from typing import Any

def print_everything(*args: Any) -> None:
    """
    Print all arguments passed to the function.

    This function accepts a variable number of positional arguments
    and prints each argument on a separate line.

    Parameters
    ----------
    *args : Any
        A variable number of positional arguments of any data type.

    Returns
    -------
    None
        This function prints the values and does not return anything.
    """
    for value in args:
        print("You entered:", value)


if __name__ == "__main__":
    """
    Example usage of the print_everything function.
    """
    print_everything("Python", "Java", "C++", "Rust")
```

```python
# 1) Without PEP

def get_top_expensive_products(products, top_n):
    # Sort products by price in descending order
    sorted_products = sorted(products, key=lambda p: p["price"], reverse=True)

    return sorted_products[:top_n]


if __name__ == "__main__":

    product_list = [
        {"name": "Laptop", "price": 90000},
        {"name": "Headphones", "price": 2000},
        {"name": "Smartphone", "price": 50000},
        {"name": "Keyboard", "price": 1500},
        {"name": "Monitor", "price": 12000},
    ]

    top_products = get_top_expensive_products(product_list, 3)

    print("Top Expensive Products:")
    for product in top_products:
        print(product)
```

```python
# 2) With PEP
from typing import List, Dict


def get_top_expensive_products(products: List[Dict], top_n: int) -> List[Dict]:
    """
    Return the top N most expensive products from a product list.

    Parameters
    ----------
    products : List[Dict]
        A list of dictionaries where each dictionary represents a product.
        Each product must contain at least the keys 'name' and 'price'.

    top_n : int
        The number of top expensive products to return.

    Returns
    -------
    List[Dict]
        A list containing the top N most expensive product dictionaries.
    """
    # Sort products by price in descending order
    sorted_products = sorted(products, key=lambda p: p["price"], reverse=True)

    return sorted_products[:top_n]


if __name__ == "__main__":

    product_list = [
        {"name": "Laptop", "price": 90000},
        {"name": "Headphones", "price": 2000},
        {"name": "Smartphone", "price": 50000},
        {"name": "Keyboard", "price": 1500},
        {"name": "Monitor", "price": 12000},
    ]

    top_products = get_top_expensive_products(product_list, 3)

    print("Top Expensive Products:")
    for product in top_products:
        print(product)
```

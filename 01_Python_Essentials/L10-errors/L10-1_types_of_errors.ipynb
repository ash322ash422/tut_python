# Types of Errors in Python: 3 main categories

- **Syntax errors**
- **Runtime errors**
- **Logical errors**

## 1. Syntax Errors

These occur when Python cannot understand your code due to incorrect syntax.

```python
# 1) SyntaxError

print("Hello)
```

```python
# 2) Fix SyntaxError

print("Hello")
```

```python
# 1) SyntaxError

if True
    print("Hello")
```

```python
# 2) Fix SyntaxError

if True:
    print("Hello")
```

```python
# 1) SyntaxError

age = 20

if age >>> 18:
    print("You are an adult")
```

```python
# 2) Fix SyntaxError

age = 20

if age > 18:
    print("You are an adult")
```

### Syntax errors happen **before** the program runs.

```python
# 2
```

## 2. Runtime Errors (Exceptions)

These occur **while** the program is running.

```python
a = 10
b = 0

print(a / b)
```

### Common runtime errors:

ZeroDivisionError

TypeError

ValueError

IndexError

KeyError

FileNotFoundError

```python
# 1) Example of TypeError
a = 10
b = "5"

result = a + b   # Cannot add int and str
print(result)
```

```python
# 2) Fix TypeError
a = 10
b = "5"

result = a + int(b) # both are int
print(result)
```

```python
# 1) Example of ValueError

age = int("twenty")   # "twenty" cannot be converted to int
print(age)
```

```python
# 2) Fix ValueError

age = int("20")   # "20" can be converted to int
print(age)
```

```python
# Example of IndexError

numbers = [77, 55, 11, 33 ]

# print(numbers[2]) 
print(numbers[5])   # Index out of range
```

```python
# Example of KeyError

student = {"name": "Rahul",
           "age": 22
          }

# print(student["name"])   # Key does exist
print(student["grade"])   # Key does not exist
```

```python
# Example of FileNotFoundError

file = open("data_file_people.txt", "r")   # File does not exist
```

## 3. Logical Errors

These are the most dangerous because:

- The program runs and **no error is shown**
- **Output is incorrect**

```python
# 1) Example: logical error

def average(a, b, c):
    return a + b + c / 3

print(average(10, 20, 30))
```

```python
# 2) Fix logical error
def average(a, b, c):
    return (a + b + c) / 3

print(average(10, 20, 30))
```

```python
# 1) Example: logical error
# Check if a number is positive
num = -5

if num > 0:
    print("Positive number")
else:
    print("Positive number")   # Logical error
```

```python
# 2) Fix logical error

num = -5

if num > 0:
    print("Positive number")
else:
    print("Negative number")  # fix
```

```python
# 1) print the total sume of list

numbers = [1, 2, 3]

total = 18   # Logical Error! (Should start at 0)

for n in numbers:
    total += n
print(total) # Output: 16 (Should be 6)
```

```python
# 2) Fix

numbers = [1, 2, 3]

total = 0   # Logical Error! (Should start at 0)

for n in numbers:
    total += n
print(total) # Output: 16 (Should be 6)
```

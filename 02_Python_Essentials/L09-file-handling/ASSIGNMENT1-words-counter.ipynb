## ASSIGNMENT: Word Counter

You are given a text file ("data_text.txt").

Analyze the text to:
- Count total words
- Find unique words
- Find longest & shortest words
- Count vowels, consonants, digits
- Find most frequent words
- Identify duplicate words

```python
# Step 1: Read the giant text file
file = open("data_text.txt", "r")
text = file.read()
file.close()

print(text)
```

```python
# Step 2: Clean and split text
text = text.lower()

for ch in ".,)\n":
    text = text.replace(ch, " ")
    
words = text.split()

print("Words:", words)
print("Total words:", len(words))
```

```python
# Step 4: Unique words (using set)
unique_words = set(words)

print("Unique words:", unique_words)
print("Count of unique words:", len(unique_words))
```

```python
# Step 5: Longest & shortest word (tuple)

longest  = words[0]
shortest = words[0]

for w in words:
    if len(w) > len(longest):
        longest = w
    elif len(w) < len(shortest):
        shortest = w

result = (longest, shortest)
print("Longest & Shortest:", result)
```

```python
# Step 6: Count vowels, consonants, digits

vowels = "aeiou"
digits = "0123456789"
consonants  = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"

vowel_count = 0
consonant_count = 0
digit_count = 0

for ch in text:
    if ch in vowels:
        vowel_count += 1
    elif ch in digits:
        digit_count += 1
    elif ch in consonants:
        consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
print("Digits:", digit_count)
```

```python
# Step 7: Word frequency 
word_count = {}

for w in words:
    word_count[w] = word_count.get(w, 0) + 1

print(word_count)
```

```python
# Step 8: Find duplicate words

word_count = {}

for w in words:
    word_count[w] = word_count.get(w, 0) + 1

duplicates = []

for word, count in word_count.items():
    if count > 1:
        print("Found duplicate:", word)
        duplicates.append(word)
        
print("Duplicates:", duplicates)
```

```python
# Create a list of words that ignores stopwords (is, the, to, and)

# step1: read the words in a list
text = text.lower()

for ch in ".,)\n":
    text = text.replace(ch, " ")
    
words = text.split()

print("Words:", words)
```

```python
# step2: Go through the list and ignore stopwords (is, the, to, and)

stopwords = ['is', 'the', 'to', 'and']
clean_words = []
for w in words:
    if w not in stopwords:
        clean_words.append(w)

print("Clean words:", clean_words)
```

```python
# Save results to a file

# step1: create a giant string from clean_words
msg = ""
for w in clean_words:
    msg = msg + " " + w
print(msg)
```

```python
# step2: write to a file
file = open("clean_junk.txt", "w")  # Opens in write mode (overwrites file)

file.write(msg) 

file.close()
```

# Assignment: 
Read a csv file "ASSIGNMENT_csv_list_dict_for.csv" and store the data as key: value pair. 

The keys should be name. The values should be stored in list and it contains age,income. 

All entries in values should be int. 

Print the content of dictionary

```python
import csv

# Path to your CSV file
file_path = "ASSIGNMENT_csv_list_dict_for.csv"

# STEP1: Read the CSV file into a list
data = []
with open(file_path, mode="r", newline="", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row)

# Print the result
print(f"read data: {data}")

print("#################################")
# STEP2: Build dictionary
result = {}
for row in data[1:]: # skip the 1st row
    print(f"processing {row}")
    result[row[0]] = row[1:]

# result = {row[0]: row[1:] for row in data[1:]}
print(result)

print("*************************************")
# STEP3: convert the values into int

for k, v in result.items():
    print(f"---key={k} and its value is {v}")
    
    temp = []
    for x in v:
        temp.append(int(x))
    result[k] = temp
    # result[k] = [int(x) for x in v]

print(f"After cleanup: {result}")
```

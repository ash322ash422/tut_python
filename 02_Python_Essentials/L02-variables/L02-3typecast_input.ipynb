# Type Casting in Python and input()  

---

## What is Type Casting?

**Type casting** means converting one data type into another.

Examples:

* Converting `int` â†’ `float`
* Converting `str` â†’ `int`
* Converting `int` â†’ `str`

Python provides **built-in functions** for type casting.

---

## ðŸ”¹ Why Do We Need Type Casting?

* User input is always a **string**.
* Performing calculations with right kind of data types 
* Avoiding runtime errors

```python
## Checking the Type of a Variable

x = 10
y = 3.5
z = "100"

print(type(x))
print(type(y))
print(type(z))
```

## Common Type Casting Functions

| Function  | Converts To |
| --------- | ----------- |
| `int()`   | Integer     |
| `float()` | Float       |
| `str()`   | String      |
| `bool()`  | Boolean     |


## Casting to Integer â†’ `int()`

```python
# step1: 

x = "10"
# print(x / 10) # ERROR
print(type(x))
```

```python
# step2: convert it into int

x = "10"
x = int(x)

print(x / 10) 
print(type(x))
```

```python
a = 10.8
x = int(a)

print(x)   # decimal part is removed
```

```python
## Invalid conversion

# This will cause an error
# int("hello")
```

```python
# 2
```

## Casting to Float â†’ `float()`

```python
a = 5   # int
x = float(a)

print(x)
```

```python
a = "3.14"  # string
x = float(a)

print(x)
```

```python
# 3
```

## Casting to String â†’ `str()`

```python
age = 25
score = 99.5

msg1 = "Age is " + str(age)
msg2 = "Score is " + str(score)

print(msg1)
print(msg2)
```

```python
# 4
```

## Casting to Boolean â†’ `bool()`

### Rules to remember:

* `False` values:

  * `0`
  * `0.0`
  * `""` (empty string)
  * `None`
* Everything else is `True`

```python
print(bool(0.0))
print(bool(1))
print(bool(""))
print(bool("Python"))
print(bool(100))
```

```python
# 5
```

# input() function
User input is **always a string**.


```python
name = input("Enter your name: ")

print("Hello,", name)
```

```python
## step1) Type Casting with User Input
age = input("Enter your age: ")
print(type(age))

# new_age = age + 1 # ERROR
```

```python
# step2) Convert it to integer:

age = int(input("Enter your age: "))

new_age = age + 1

print("Next year your age will be:", new_age)
```

```python
## Real-Life Example: Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum_result = num1 + num2
print("Sum:", sum_result)
```

## Write a function that returns mean, min, max of a list

```python

def mean_min_max(data):
    total = 0
    minimum = data[0]
    maximum = data[0]

    for x in data:
        total += x
        if x < minimum:
            minimum = x
        if x > maximum:
            maximum = x

    mean = total / len(data)
    return mean, minimum, maximum
```

```python
age = [10, 21, 12, 9, 67]
result = mean_min_max(age)
print(result)
```

## Write a function that gives Frequency Distribution of whole numbers

```python
def frequency_distribution(data):
    freq = {}

    for x in data:
        if x in freq:
            freq[x] += 1
        else:
            freq[x] = 1

    return freq
```

```python
age = [10, 21, 10, 21, 10, 34, 10, 34]
result = frequency_distribution(age)
print(result)
```

```python
def separate_even_odd(data):
    even = []
    odd = []

    for x in data:
        if x % 2 == 0:
            even.append(x)
        else:
            odd.append(x)

    return even, odd
```

```python
# Median (Sorted Data)
def median(data):
    n = len(data)

    if n % 2 == 1:
        return data[n // 2]
    else:
        return (data[n // 2 - 1] + data[n // 2]) / 2
```

```python
# Count Values Above Mean
def count_above_mean(data):
    total = 0
    for x in data:
        total += x

    mean = total / len(data)

    count = 0
    for x in data:
        if x > mean:
            count += 1

    return count
```

```python
# Variance
def variance(data):
    total = 0
    for x in data:
        total += x

    mean = total / len(data)

    variance_sum = 0
    for x in data:
        variance_sum += (x - mean) ** 2

    return variance_sum / len(data)
```

```python
# Categorize Using While Loop
def categorize_numbers(data):
    result = {"<=50": [], ">50": []}

    i = 0
    while i < len(data):
        if data[i] <= 50:
            result["<=50"].append(data[i])
        else:
            result[">50"].append(data[i])
        i += 1

    return result
```

```python
# Mode of Dataset
def mode(data):
    freq = {}

    for x in data:
        if x in freq:
            freq[x] += 1
        else:
            freq[x] = 1

    mode_value = None
    max_count = 0

    for key in freq:
        if freq[key] > max_count:
            max_count = freq[key]
            mode_value = key

    return mode_value
```

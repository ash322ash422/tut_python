## Problem Statement

You are given an email address as input.
Write a program that extracts:

Username

Domain name

Email provider (gmail, yahoo, etc.)

### Example Input
```
john.doe@gmail.com
```


### Expected Output
```
Username: john.doe
Domain: gmail.com
Provider: gmail
```


````python
# SOLUTION: Email Parser Program

email = input("Enter email address: ")

# Split email into username and domain
username, domain = email.split("@")

# Extract provider name
provider = domain.split(".")[0]

# Display results using f-strings
print(f"Username  : {username}")
print(f"Domain    : {domain}")
print(f"Provider  : {provider}")
````

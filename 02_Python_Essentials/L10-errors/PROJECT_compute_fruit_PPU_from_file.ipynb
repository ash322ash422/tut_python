# Mini Project: Price Per Unit Calculator (CSV + Error Handling)

### Project Objective

Read a CSV file containing: fruit, total_price, qty

Calculate price per unit

Handle file and data errors

Continue processing even if some rows are invalid

### The content of file:
```
fruit,total_price,qty
Kiwi,120,6
Guava,30,3
Banana,50,0
Orange,abc,4
Mango,200,5
```

````python
import csv

file_name = "fruits.csv"

try:
    with open(file_name, "r") as file:
        reader = csv.DictReader(file)

        for row in reader:
            try:
                fruit = row["fruit"]
                total_price = float(row["total_price"])
                qty = int(row["qty"])

                price_per_unit = total_price / qty
                print(f"Fruit: {fruit} | Price per unit: {price_per_unit}")

            except ZeroDivisionError:
                print(f"Error: Cannot calculate price for {fruit} (Quantity is zero)")

            except ValueError:
                print(f"Error: Invalid data for {row.get('fruit', 'Unknown')}")

except FileNotFoundError:
    print("ERROR: CSV file not found. Please check the file name.")

finally:
    print("ERROR: Processing completed.")
````

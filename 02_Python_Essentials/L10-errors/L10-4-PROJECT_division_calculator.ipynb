## Project Objective

Create a simple Division Calculator that:

Takes input from the user

Performs division

Handles common errors gracefully

Does not crash the program

### Project Requirements

The calculator should:

1. Ask the user for two numbers

2. Divide the first number by the second

3. Handle: "Division by zero" and "Invalid input (non-numeric)"

5. Ask the user if they want to continue or exit

```python
## Concepts Covered

input()

Type conversion (int, float)

try-except

Catching multiple errors

else and finally

Looping for repeated use
```

# Solution

```python
# step1: Basic Division (Without Error Handling)

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

result = a / b
print("Result:", result)
```

```python
# Step 2: Division Calculator with Error Handling
while True:
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        result = num1 / num2

    except ValueError:
        print("Error: Invalid input. Please enter numbers only.")

    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    else:
        print("Result:", result)

    finally:
        choice = input("Do you want to continue? (yes/no): ").lower()
        if choice != "yes":
            print("Calculator closed.")
            break
```

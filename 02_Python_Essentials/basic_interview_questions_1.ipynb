### 3. What’s the difference between list, tuple, and set?

Answer:

List: Ordered, mutable, allows duplicates → ["a", "b"]

Tuple: Ordered, immutable, allows duplicates → ("a", "b")

Set: Unordered, mutable, no duplicates → {"a", "b"}



## 4. What is the difference between is and == in Python?

Answer:

== -> compares values (content).

is -> compares object identity (memory location).


```python
a = [1, 2]
b = [1, 2]
print(a == b)  # True (values are equal)
print(a is b)  # False (different objects in memory)
```

### 5. What are Python’s mutable and immutable types?

Answer:

Mutable: Can be changed after creation → list, dict, set.

Immutable: Cannot be changed → int, float, str, tuple, frozenset.

### 8. What are Python’s conditional statements?

Answer:

if, elif, else.

```python
x = 10
if x > 0:
    print("Positive")
elif x == 0:
    print("Zero")
else:
    print("Negative")
```

### 9. What are Python’s loops?

Answer:

for loop → iterates over sequences.

while loop → runs until condition is false.

```python
for i in range(3):
    print(i)

x = 0
while x < 3:
    print(x)
    x += 1
```

### 10. What are Python functions and how are they defined?

Answer:
Functions are reusable blocks of code defined using def.

```python
def greet(name="Guest"):
    return f"Hello, {name}"

print(greet("Friend"))  
print(greet())          
```

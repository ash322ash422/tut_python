```python
# Problem Without Dictionary Comprehension
# Create a dictionary of student -> marks

students = ["Ash", "Sarah", "John"]
marks = [85, 92, 78]

student_marks = {}

for i in range(len(students)):
    student_marks[students[i]] = marks[i]

print(student_marks)
```

```python
# 1) Dictionary comprehension syntax:
# {key: value for item in iterable}

numbers = [1, 2, 3, 4, 5]

square_dict = {n: n**2 for n in numbers}

print(square_dict)
```

### Equivalent to loop but shorter and cleaner.

```python
# 2) Student Marks Mapping
students = ["Amit", "Sara", "John"]
marks = [85, 92, 78]

student_marks = {students[i]: marks[i] for i in range(len(students))}

print(student_marks)
```

# Using Conditions in Dictionary Comprehension

```python
# Keep only students scoring above 80

student_marks = {
    "Amit": 85,
    "Sara": 92,
    "John": 78
}

high_scorers = {k: v for k, v in student_marks.items() if v > 80}

print(high_scorers)
```

# Modify Values Using Comprehension

```python
# Add bonus marks (+5)

updated_marks = {k: v + 5 for k, v in student_marks.items()}

print(updated_marks)
```

# A real world example:

```python
# 1) Product data (like from an API)

products = [
    {"name": "Laptop", "price": 60000, "stock": 10},
    {"name": "Mouse",  "price": 500, "stock": 50},
    {"name": "Keyboard", "price": 1500, "stock": 20}
]

print(products)
```

```python
# 2) Convert List → Structured Lookup Dictionary

# Create product lookup dictionary
# product_name -> price

product_price = {p["name"]: p["price"] for p in products}

print(product_price)

# Fast lookup system (used in real applications).
```

```python
# 3) Calculate total inventory value per product
# price * stock

inventory_value = {
    p["name"]: p["price"] * p["stock"]
    for p in products
}

print(inventory_value)
```

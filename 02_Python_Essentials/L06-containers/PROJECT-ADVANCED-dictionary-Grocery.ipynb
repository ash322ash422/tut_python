Here students will:
- Work with lists and dictionaries together
- Use loops for menu-driven programs
- Apply nested conditionals
- Perform CRUD operations (Create, Read, Update, Delete)


### You are given following dictionary:
```
grocery = {
    "rice": {"quantity": 2, "price": 60},
    "milk": {"quantity": 1, "price": 50}
}
```

### Your task is to implement following features:
```
1. Add item
2. View all items
3. Update item
4. Remove item
5. Search item
6. Calculate total bill
7. Exit
```


## 1. Add Grocery Item

Ask for:

Item name

Quantity

Price per unit

If item already exists → show warning

Else → add to dictionary



## 2. View Grocery List

Display

```
Item     Quantity     Price     Total
--------------------------------------
Rice        2          60        120
Milk        1          50         50
```



## 3. Update an Item

User chooses item name

Options:

Update quantity

Update price

Handle item-not-found case



## 4. Remove an Item

Remove item by name

Show confirmation message



## 5. Search for an Item

Search by item name

If found → show details

Else → show "Item not found"



## 6. Calculate Total Grocery Bill

Loop through items

Multiply quantity × price

Display final bill amount



## 7. Exit Program

Clean exit using break

# SOLUTION

````python
# Grocery List Manager

grocery = {}

while True:
    print("\n--- Grocery List Manager ---")
    print("1. Add item")
    print("2. View items")
    print("3. Update item")
    print("4. Remove item")
    print("5. Search item")
    print("6. Calculate total bill")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    # 1. Add Item
    if choice == "1":
        item = input("Enter item name: ").lower()

        if item in grocery:
            print("Item already exists!")
        else:
            quantity = int(input("Enter quantity: "))
            price = float(input("Enter price per unit: "))

            grocery[item] = {
                "quantity": quantity,
                "price": price
            }
            print("Item added successfully!")

    # 2. View Items
    elif choice == "2":
        if not grocery:
            print("Grocery list is empty.")
        else:
            print("\nItem\tQuantity\tPrice\tTotal")
            print("-" * 40)
            for item, details in grocery.items():
                total = details["quantity"] * details["price"]
                print(f"{item}\t{details['quantity']}\t\t{details['price']}\t{total}")

    # 3. Update Item
    elif choice == "3":
        item = input("Enter item name to update: ").lower()

        if item in grocery:
            print("1. Update quantity")
            print("2. Update price")

            update_choice = input("Enter choice (1/2): ")

            if update_choice == "1":
                new_quantity = int(input("Enter new quantity: "))
                grocery[item]["quantity"] = new_quantity
                print("Quantity updated successfully!")

            elif update_choice == "2":
                new_price = float(input("Enter new price: "))
                grocery[item]["price"] = new_price
                print("Price updated successfully!")

            else:
                print("Invalid update option.")
        else:
            print("Item not found.")

    # 4. Remove Item
    elif choice == "4":
        item = input("Enter item name to remove: ").lower()

        if item in grocery:
            del grocery[item]
            print("Item removed successfully!")
        else:
            print("Item not found.")

    # 5. Search Item
    elif choice == "5":
        item = input("Enter item name to search: ").lower()

        if item in grocery:
            details = grocery[item]
            total = details["quantity"] * details["price"]
            print(f"Item: {item}")
            print(f"Quantity: {details['quantity']}")
            print(f"Price: {details['price']}")
            print(f"Total: {total}")
        else:
            print("Item not found.")

    # 6. Calculate Total Bill
    elif choice == "6":
        total_bill = 0

        for details in grocery.values():
            total_bill += details["quantity"] * details["price"]

        print(f"Total Grocery Bill: ₹{total_bill}")

    # 7. Exit
    elif choice == "7":
        print("Thank you for using Grocery List Manager!")
        break

    else:
        print("Invalid choice. Please select between 1 and 7.")
````

# ASSIGNMENT: Shopping cart
You are building a simple Shopping Cart system.

1. Create a shopping cart with the following items:
   "milk", "bread", "eggs"

2. Display the first item and the last item in the cart.

3. Add "butter" and "cheese" to the cart.

4. Remove "bread" from the cart.

5. Display only the first three items using slicing.

6. Convert the final shopping cart into a tuple
   to prevent further changes (make it immutable).

7. Try modifying the tuple and observe what happens.


```python
# Step 1: Create the shopping cart (list)
shopping_cart = ["milk", "bread", "eggs"]
print("Initial shopping cart:", shopping_cart)
```

```python
# Step 2: Access items using indexing
print("First item in cart:", shopping_cart[0])
print("Last item in cart:", shopping_cart[-1])
```

```python
# Step 3: Add new items using append()
shopping_cart.append("butter")
shopping_cart.append("cheese")
print("Cart after adding items:", shopping_cart)
```

```python
# Step 4: Remove an item using remove()
shopping_cart.remove("bread")
print("Cart after removing bread:", shopping_cart)
```

```python
# Step 5: Display first three items using slicing
print("First three items:", shopping_cart[:3])
```

```python
# Step 6: Convert list to tuple (to make it immutable)
final_cart = tuple(shopping_cart)
print("Final cart (tuple):", final_cart)
```

# Explanation:

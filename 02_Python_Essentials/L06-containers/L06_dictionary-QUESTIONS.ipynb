## Dictionary Keys as a Set
Question

Given a dictionary of student scores, extract all unique subjects.

```python
scores = {
    "Alice": ("Math", "Science"),
    "Bob": ("Science", "English"),
    "Cara": ("Math", "English")
}
```

```python
result = set(scores["Alice"]) | set(scores["Bob"]) | set(scores["Cara"])
```

## Common Values Between Dictionary Entries
Question

Find subjects common to all students.

```python
scores = {
    "Ahmed": ("Math", "Science"),
    "Happy": ("Science", "English"),
    "Cara": ("Math", "English")
}
```

```python
result = set(scores["Ahmed"]) & set(scores["Happy"]) & set(scores["Cara"])
```

## Create Dictionary from Tuple
Question

Create a dictionary by pairing keys and values from two tuples.

```python
keys = ("Math", "Science", "English")
values = (90, 85, 88)
```

```python
result = dict(zip(keys, values))
```

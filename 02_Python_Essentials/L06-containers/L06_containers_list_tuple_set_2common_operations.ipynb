# Common operations on list, tuple, set
- **len(), max(), min(), sum()**
- **in**  membership operator

```python
# numbers = [10, 5, 20, 8] # list
numbers = (10, 5, 20, 8) # tuple

print("Length  :", len(numbers))
print("Maximum :", max(numbers))
print("Minimum :", min(numbers))
print("Sum     :", sum(numbers))
```

```python
# average
numbers = [10, 5, 20, 8]

avg = sum(numbers) / len(numbers)

print("Average:", avg)
```

# Membership Test

```python
numbers = [10, 5, 20, 8] # list

b = 10 in numbers
print(b)
```

```python
numbers = (10, 5, 20, 8) # tuple

b = 10 in numbers
print(b)
```

```python
numbers = {10, 5, 20, 8} # set

b = 10 in numbers
print(b)
```

## ASSIGNMENT

Write a program that takes sentence input from a user and performs following:

Counts total words

Counts total characters

Finds the most frequent word


### Example Input
```
Data science is fun and data science is powerful
```


### Expected Output
```
Total Words: 8
Total Characters: 46
Most Frequent Word: data
```

````python
# SOLUTION: Word Counter Program

text = input("Enter text: ")

# Convert text to lowercase for consistency
text = text.lower()

# Split text into words
words = text.split()

# Count total words
total_words = len(words)

# Count total characters (excluding spaces)
total_characters = len(text.replace(" ", ""))

# Word frequency dictionary
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Find most frequent word (beginner-friendly approach)
max_count = 0
most_frequent_word = ""

for word in word_count:
    if word_count[word] > max_count:
        max_count = word_count[word]
        most_frequent_word = word


# Display results using f-strings
print(f"Total Words       : {total_words}")
print(f"Total Characters  : {total_characters}")
print(f"Most Frequent Word: {most_frequent_word}")
````

```python
# Advanced quiz (no random, no string, no enumerate)


QUESTIONS = {
    "Which planet is known as the Red Planet?": [
        "mars", "jupiter", "venus", "mercury",
    ],
    "Which planet is the largest in our solar system?": [
        "jupiter", "saturn", "neptune", "earth",
    ],
}

LABELS = ['a', 'b', 'c', 'd']

num_questions = len(QUESTIONS)

num_correct = 0
question_number = 1

for question in QUESTIONS:
    if question_number > num_questions:
        break

    print(f"\nQuestion {question_number}:")
    print(f"{question}?")

    alternatives = QUESTIONS[question]
    correct_answer = alternatives[0]

    labeled_alternatives = dict(zip(LABELS, alternatives))

    for label in labeled_alternatives:
        print(f"  {label}) {labeled_alternatives[label]}")

    while True:
        answer_label = input("\nChoice? ").lower()
        if answer_label in labeled_alternatives:
            break
        print(f"Please answer one of {', '.join(labeled_alternatives)}")

    answer = labeled_alternatives[answer_label]

    if answer == correct_answer:
        num_correct += 1
        print("Correct!")
    else:
        print(f"The answer is {correct_answer!r}, not {answer!r}")

    question_number += 1

print(f"\nYou got {num_correct} correct out of {question_number - 1} questions")
```

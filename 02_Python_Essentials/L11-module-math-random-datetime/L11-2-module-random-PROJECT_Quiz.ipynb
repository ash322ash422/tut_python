```python
import random
```

## Create Quiz Question Bank
Each question is stored as a dictionary

All questions stored in a list

```python
quiz_questions = [
    {
        "question": "What is the output of: print(2 ** 3)?",
        "options": ["6", "8", "9", "5"],
        "answer": "8"
    },
    {
        "question": "Which keyword is used to define a function in Python?",
        "options": ["func", "define", "def", "function"],
        "answer": "def"
    },
    {
        "question": "Which data type is immutable?",
        "options": ["List", "Dictionary", "Set", "Tuple"],
        "answer": "Tuple"
    },
    {
        "question": "What does len([1,2,3]) return?",
        "options": ["2", "3", "Error", "None"],
        "answer": "3"
    },
    {
        "question": "Which module is used to generate random numbers?",
        "options": ["math", "random", "datetime", "os"],
        "answer": "random"
    }
]
```

## Randomly Select Questions

```python
def get_random_questions(question_bank, num_questions):
    return random.sample(question_bank, num_questions)
```

## Shuffle Options Randomly

```python
def ask_question(question_data):
    print("\n" + question_data["question"])

    options = question_data["options"].copy()
    random.shuffle(options)

    for idx, option in enumerate(options, start=1):
        print(f"{idx}. {option}")

    user_choice = int(input("Enter your choice (1-4): "))
    selected_option = options[user_choice - 1]

    return selected_option == question_data["answer"]
```

random.shuffle() changes order in-place

.copy() avoids modifying original data

## Build the Quiz Engine

```python
def run_quiz():
    score = 0
    total_questions = 3

    selected_questions = get_random_questions(quiz_questions, total_questions)

    for q in selected_questions:
        if ask_question(q):
            print("Correct!")
            score += 1
        else:
            print("Wrong!")

    print("\nQuiz Completed!")
    print(f"Your Score: {score}/{total_questions}")
```

## Run the Project

```python
run_quiz()
```

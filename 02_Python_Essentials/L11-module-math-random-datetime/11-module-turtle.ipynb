## ğŸ¢ Create a Turtle Screen & Turtle

```python
screen = turtle.Screen()
screen.bgcolor("black")

pen = turtle.Turtle()
pen.shape("turtle")
pen.color("cyan")
pen.speed(3)
```

---

## âœï¸ Basic Drawing: A Square

```python
for _ in range(4):
    pen.forward(100)
    pen.right(90)
```

---

## ğŸ”º Drawing a Triangle (Using Loop)

```python
pen.clear()
pen.color("yellow")

for _ in range(3):
    pen.forward(120)
    pen.left(120)
```

---

## ğŸŒ¸ Drawing a Colorful Flower (Interesting Part!)

```python
pen.clear()
pen.speed(0)

colors = ["red", "purple", "blue", "green", "orange", "yellow"]

for i in range(36):
    pen.color(colors[i % 6])
    pen.circle(100)
    pen.left(10)
```

* Loops
* Colors
* Rotation
* Circles

## ğŸ¯ Mini Fun Example: Spiral Design

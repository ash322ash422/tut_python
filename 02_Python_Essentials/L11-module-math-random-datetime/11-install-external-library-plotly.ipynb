## ğŸ“Š Example 1: Interactive Line Plot (Very Cool)

```python
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Visitors": [120, 150, 180, 170, 220, 300, 280]
}

df = pd.DataFrame(data)

fig = px.line(
    df,
    x="Day",
    y="Visitors",
    title="ğŸ“ˆ Website Visitors Over a Week",
    markers=True
)

fig.show()
```

## ğŸŒˆ Example 2: Colorful Bar Chart

```python
fig = px.bar(
    df,
    x="Day",
    y="Visitors",
    color="Visitors",
    title="ğŸ“Š Visitors Per Day (Color Mapped)",
    text="Visitors"
)

fig.show()
```

## ğŸ«§ Example 3: Bubble Chart (Eye-Catching)

## ğŸŒ Example 4: Animated Scatter Plot (WOW Factor)

# Python `datetime` Module (Beginner Notebook)

## What is the `datetime` Module?

The **`datetime` module** is a **built-in Python library** used to work with:

* Dates
* Time
* Date + Time together

Common use cases:

* Timestamps
* Logging
* Age calculators
* Attendance systems
* Time-based programs

---


```python
## Importing `datetime`
import datetime
```

## Getting Todayâ€™s Date

```python
today = datetime.date.today()
print(today)
```

## Getting Current Date and Time


```python
now = datetime.datetime.now()
print(now)
```

## Accessing Date & Time Parts

```python
print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)

print("Hour:", now.hour)
print("Minute:", now.minute)
print("Second:", now.second)
```

## Date Difference (Timedelta)

```python
### Days Between Two Dates

start_date = datetime.date(2025, 1, 1)
end_date = datetime.date(2025, 1, 10)

difference = end_date - start_date
print(difference)
```

```python
### Access Number of Days

print(difference.days)
```

## Adding/Subtracting Days to a Date

```python
today = datetime.date.today()
print(today)

d = today + datetime.timedelta(days=7) # or use -
# d = today + datetime.timedelta(hours=700) # or use -

print(d)
```

## Formatting Date & Time (Very Important)

```python
### Convert Date to Readable String

print(now.strftime("%Y-%m-%d"))
```

### Common Format Codes

| Code | Meaning    |
| ---- | ---------- |
| `%Y` | Year       |
| `%m` | Month      |
| `%d` | Day        |
| `%H` | Hour (24h) |
| `%M` | Minute     |
| `%S` | Second     |


```python
### Example Format

print(now.strftime("%d-%m-%Y %H:%M:%S"))
```

## Convert String to Date

```python
date_string = "25-12-2025"
converted_date = datetime.datetime.strptime(date_string, "%d-%m-%Y")
print(converted_date)
```

## Mini Example: Age Calculator

```python
birth_year = int(input("Enter your birth year: "))
current_year = datetime.date.today().year

age = current_year - birth_year
print("Your age is:", age)
```

```python
# following prints good morning/afternoon/evening depending when u run this
hr = datetime.datetime.now().hour

if hr < 12:
    print("Good morning!")
elif hr < 18:
    print("Good afternoon!")
else:
    print("Good evening!")
```

```python
# Application: adding a timestamp on log files
now = datetime.datetime.now()
filename = now.strftime("logfile_%Y%m%d_%H%M%S.txt")


# some data is written to above filename
with open(filename, 'w') as file:
    t = now.strftime("%m/%d/%Y, %H:%M:%S")
    file.write(f"{t}: caught SIGTERM, shutting down ")# Write the message to the file
    
print("Saving to:", filename)
```

# Python `random` Module (Beginner Essentials)

## What is the `random` Module?
The **`random` module** is a **built-in Python library** used to generate **random numbers and choices**.

It is commonly used in:

* Games üéÆ
* Simulations
* Sampling data
* Simple AI / ML demos
* Password & OTP examples (basic level)



````python
import random
````

## 1. Generate a Random Integer

### `random.randint(start, end)`

Returns a **random whole number** between `start` and `end` (both included).


````python
# Example use: Dice roll

print(random.randint(1, 6))
````

## 2Ô∏è. Generate a Random Float

### `random.random()`

Returns a **random decimal number** between `0.0` and `1.0`

````python
# Example use: Probability, simulations

print(random.random())
````

### `random.uniform(a, b)`

Returns a **random decimal number** between `a` and `b`

````python
print(random.uniform(10, 20))
````

## 3. Pick a Random Item from a List

### `random.choice(list)`

````python
# Example use: Random quiz question, game character

fruits = ["guava", "apple", "cherry", "mango"]
print(random.choice(fruits))
````

## 4Ô∏è. Shuffle a List

### `random.shuffle(list)`

> This **changes the original list**


````python
# Example use: Card games, random order of questions

numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)
````

## 5Ô∏è. Pick Multiple Random Items

### `random.sample(list, count)`


````python
students = ["A", "B", "C", "D"]
print(random.sample(students, 2))
print(students)
````

###  No repetition
### Original list is unchanged

## Summary
```
randint() ‚Üí random integer
random()  ‚Üí decimal between 0 and 1
choice()  ‚Üí pick one item
shuffle() ‚Üí mix a list
sample()  ‚Üí pick multiple items
```

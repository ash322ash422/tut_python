# 5. OPERATORS – TOTAL, AVERAGE, TOPPER

##  Exercise 5

For each student:

1. Calculate:

   * Total marks
   * Average marks
2. Find:

   * Class topper

---

````python
print(students)
````

````python
### Solution 5

topper_name = ""
max_total = 0

for s in students[1:]:
    marks = [ int(s[3]), int(s[4]), int(s[5]) ]
    total = sum(marks)
    average = total / 3

    print(s[1], "Total:", total, "Average:", average)

    if total > max_total:
        max_total = total
        topper_name = s[1]

print("\nTopper:", topper_name, "with", max_total)
````

# 6. IF–ELSE – PASS / FAIL & GRADE

## Exercise 6

Using average:

* ≥ 90 → A
* ≥ 75 → B
* ≥ 60 → C
* Else → Fail

---



````python
### Solution 6

for s in students[1:]:
    marks = [ int(s[3]), int(s[4]), int(s[5]) ]
    avg = sum(marks) / 3

    if avg >= 90:
        grade = "A"
    elif avg >= 75:
        grade = "B"
    elif avg >= 60:
        grade = "C"
    else:
        grade = "Fail"

    print(s[1], "Grade:", grade)
````

# 8. TODO(TEST): WHILE LOOP – MENU-DRIVEN PROGRAM

## Exercise 8

Create a menu:

```
1. Show All Students
2. Search by Roll No
3. Show Topper
4. Show Failed Students
5. Show Unique Classes
6. Exit
```

---



````python
### Solution 8 (Complete Menu Program)


import csv

students = []

with open("data_student_result.csv", "r") as file:
    reader = csv.reader(file)
    
    for row in reader:
        students.append(row)

print("The total students:", len(students)-1)

while True:
    print("\n1. Show All Students")
    print("2. Search by Roll No")
    print("3. Show Topper")
    print("4. Show Failed Students")
    print("5. Show Unique Classes")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        for s in students:
            print(s)

    elif choice == "2":
        r = input("Enter Roll No: ")
        found = False
        for s in students[1:]:
            if s[0] == r:
                print(s)
                found = True
        if not found:
            print("Student not found")

    elif choice == "3":
        topper = ""
        max_total = 0

        for s in students[1:]:
            marks = [ int(s[3]), int(s[4]), int(s[5]) ]
            total = sum(marks)

            if total > max_total:
                max_total = total
                topper = s[1]

        print("Topper:", topper, "Total:", max_total)

    elif choice == "4":
        print("Failed Students:")
        for s in students[1:]:
            marks = [ int(s[3]), int(s[4]), int(s[5]) ]
            avg = sum(marks) / 3

            if avg < 60:
                print(s[1], "Avg:", avg)

    elif choice == "5":
        classes = set()
        for s in students[1:]:
            classes.add(s[2])
        print("Unique Classes:", classes)

    elif choice == "6":
        print("Exiting Program...")
        break

    else:
        print("Invalid choice")
````

1) What Will Be the Output? (Dynamic Typing)
x = 10
x = "Python"
print(x)


Options:
A. 10
B. "Python"
C. Error: type mismatch
D. None

Correct Answer: B

Python is dynamically typed. A variable can change its data type during execution.

2) Which of the following variable names is NOT valid in Python?

A. _total_marks
B. student2
C. total-marks
D. marks_2026

Correct Answer: C

Hyphens (-) are not allowed in variable names because Python treats them as subtraction operators.

# PROJECT: AGE CALCULATOR

The user enters his year of birth, then program calculates a person's age based on their year of birth. 

Concepts covered:

- Variables
- input()
- Typecasting (int)
- Arithmetic operations
- f-strings

```python
# ASK USER FOR INPUT
# input() always returns a string

name = input("Enter your name: ")
birth_year = input("Enter your year of birth: ")

# TYPECASTING: Convert birth year from string to integer
birth_year = int(birth_year)

# CURRENT YEAR: We store the current year in a variable
current_year = 2026

# Age = Current Year - Year of Birth
age = current_year - birth_year

# DISPLAY OUTPUT USING F-STRINGS
print("----- AGE CALCULATOR RESULT -----")
print(f"Name: {name}")
print(f"Year of Birth: {birth_year}")
print(f"Current Year: {current_year}")
print(f"Your Age is: {age} years")
```

```python
# (EXTRA CONCEPT) BOOLEAN LOGIC 
# Check if the user is an adult

is_adult = age >= 18

print(f"Adult Status: {is_adult}")

# END OF PROGRAM
```

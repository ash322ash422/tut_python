## Mini Project: Temperature converter (Celsius to Fahrenheit)

Write a program that takes input in Celsius from user, converts it to Fahrenheit, and print the result.
The output should be something like below

- fahrenheit = (celsius * 9/5) + 32

```
Temperature Converter
Celsius to Fahrenheit
Enter the temperature (in Celsius):  20
Temperature in Fahrenheit: 68

```


# Solution

````python
print("Temperature Converter")
print("Celsius to Fahrenheit")

celsius = float(input("Enter the temperature (in Celsius): "))

fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit:", fahrenheit)
````

# Mean, Median, Variance, Std Dev (from scratch)

```python
data = [10, 12, 15, 18, 20, 22, 25]

count = len(data)
total = sum(data)
mean = total / count

print("mean:", mean)
```

```python
# median

data = [10, 12, 15, 18, 19, 20, 22, 25]
count = len(data)
print(count // 2)

sorted_data = sorted(data)
if count % 2 != 0:
    median = sorted_data[count // 2]
else: 
    mid1 = sorted_data[count // 2]
    mid2 = sorted_data[count // 2 + 1]
    median = (mid1 + mid2)/2

print(median)
```

```python
# variance and std dev
data = [10, 12, 15, 18, 20, 22, 25]

count = len(data)
total = sum(data)
mean = total / count

numerator = []
for x in data:
    res = (x - mean)**2
    numerator.append(res)

sum_numerator = sum(numerator)
variance = sum_numerator / count
print("Variance", variance)

std_dev = variance ** 0.5
print("Standard deviation:", std_dev)
```

# Distance Metrics (VERY IMPORTANT for ML)

Matters in following

- KNN
- Clustering
- Similarity search

```python
x1 = [2, 3, 4]
x2 = [5, 7, 9]

distance = 0
for i in range(len(x1)):
    distance += (x1[i] - x2[i]) ** 2

distance = distance ** 0.5
print("Euclidean Distance:", distance)
```

# Normalization (Feature Scaling)
Matters in

- Gradient descent
- Neural networks
- Distance-based models

```python
data = [50, 60, 70, 80, 90]

min_val = min(data)
max_val = max(data)

scaled = []
for x in data:
    scaled.append((x - min_val) / (max_val - min_val))

print(scaled)
```

# Probability Basics (CORE of ML & AI)
matters

- Naive Bayes
- Model confidence
- Risk estimation

```python
labels = ['yes', 'no', 'yes', 'yes', 'no', 'yes']

total = len(labels)
yes_count = labels.count('yes')

prob_yes = yes_count / total
print("P(Yes):", prob_yes)
```

# Linear Equation (Regression Foundation)
matters

- Linear Regression
- Loss functions

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 9, 11]  # perfect linear

m = (y[-1] - y[0]) / (x[-1] - x[0])
b = y[0] - m * x[0]

print("Slope:", m)
print("Intercept:", b)
```

```python
# Error Metrics (MODEL EVALUATION)
matters

- Model comparison
- Optimization
```

```python
y_true = [3, 5, 7, 9]
y_pred = [2.5, 5.1, 6.8, 9.2]

error = 0
for i in range(len(y_true)):
    error += abs(y_true[i] - y_pred[i])

mae = error / len(y_true)
print("MAE:", mae)
```

```python
# Vector Dot Product (Neural Networks!)
matters

- Neural networks
- Linear models
```

```python
weights = [0.2, 0.4, 0.6]
inputs =  [1.0, 2.0, 3.0]

dot = 0
for i in range(len(weights)):
    dot += weights[i] * inputs[i]

print("Dot Product:", dot)
```

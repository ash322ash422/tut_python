### 2) Create a simple CLI program cli_even_odd.py that takes 2 input from user, then adds them and prints result is odd or even
- Run it in Command Line Interface (CLI)

> python cli_even_odd.py


```python
num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))

result = num1 + num2

if result % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### 2) Create a simple CLI program cli_even_odd2.py that takes 2 input from user, then adds them and prints result is odd or even. This time use an entry point __main__:

- Run it in Command Line Interface (CLI)

> python cli_even_odd2.py


```python

if __name__ == '__main__':
       
    num1 = int(input("Enter num1: "))
    num2 = int(input("Enter num2: "))
    
    result = num1 + num2
    
    if result % 2 == 0:
        print("Even")
    else:
        print("Odd")
```

### 3) Create a simple CLI program cli_multiply5.py that takes numeric input from user, then multiply it by 5 and prints result. If the user enters 0, the program quits. Use an entry point __main__:


- Run it in Command Line Interface (CLI)

> python cli_multiply5.py


```python
#  while Loop with User Input

if __name__ == '__main__':
        
    num = int(input("Enter a number to multiply by 5 (0 to stop): "))
    while num != 0:
        print(f"{num} multiply by 5: {num*5}")
        num = int(input("Enter a number to multiply by 5 (0 to stop): "))
    
    print("Good Bye")
```

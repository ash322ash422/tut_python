# Nested if statement

Example:
```
if [boolean_expression1]:
  if [boolean_expression2]:
      # block of code
  else:
      # block of code
elif [boolean_expression3]:
  if [boolean_expression4]:
      # block of code
  else:
      # block of code
else:
      # block of code
```

````python
# If age > 18 and has license, then eligible to drive
# if age > 18 and no license, then you are over 18 but need Driver license to be on road
# if age < 18, then not eligible to drive

age = 19 # change to 12, 19, 
has_license = False # True, False

if age >= 18:
    if has_license:
        print("You are eligible to drive.")
    else:
        print("You are over 18, but you need a driving license.")

else:
    print("You are younger than 18. You are too young to drive. ")
````

````python
# if attendance >= 75 and fees is paid, then student is eligible for exam
# if attendance >= 75 and fees is NOT paid, then student needs to pay fees to appear for exam
# if attendance < 75, then student is falling low on attendance

attendance = 80 # change to 60, 80
fees_paid = False # change to True, False

if attendance >= 75:
    if fees_paid:
        print("Eligible for exam")
    else:
        print("Pay fees to appear for exam")
else:
    print("Attendance shortage")
````

## A bank decides loan approval based on:

### Level 1 Decision
- Check credit score (Should be > 700)

### Level 2 Decisions
- If credit score is good → Check income (income should be > 50000)
- If credit score is bad → Check collateral

````python
credit_score = 720
monthly_income = 55000
has_collateral = False

if credit_score >= 700:
    print("Good Credit Score")

    # Nested IF inside IF
    if monthly_income >= 50000:
        print("Loan Approved")
    else:
        print("Loan Approved with Lower Amount")

else:
    print("Low Credit Score")

    # Nested IF inside ELSE
    if has_collateral:
        print("Loan Approved Against Collateral")
    else:
        print("Loan Rejected")
````

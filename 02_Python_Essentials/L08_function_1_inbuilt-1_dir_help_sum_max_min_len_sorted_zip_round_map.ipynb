## sum(), min(), max(), sorted()

```python
numbers = [3, 7, 5, -1] # list 
# numbers = (3, 7, 5, -1) # tuple

res = sum(numbers)
print("Sum:", res)
```

```python
res = min(numbers)
print("Minimum:", res)
```

```python
res = max(numbers)
print("Maximum:", res)
```

```python
res = sorted(numbers)
print("Sorted numbers:", res)
```

### len()
This gives the length of list, string, tuple, set, dictionary,etc

```python
# Need for len function:
# I want to count the num of elements in containers for data

fruits = [ "Guava", "Kiwi", "Mango", "Kiwi" ] # list
# fruits = ( "Guava", "Kiwi", "Mango", "Kiwi" ) # tuple

print(len(fruits))
```

```python
fruits = { "Guava", "Kiwi", "Mango" } # set
print(len(fruits))
```

```python
fruits_dict = { "Guava": 2,
                "Kiwi": 4 } # dictionary. Fruits:quantity

print(len(fruits_dict))
```

## round()

## map() — Feature Transformation

```python
data = ["1", "2", "3"]
print(data)

numbers = list(map(int, data)) # convert all number to int data type
print(numbers)
```

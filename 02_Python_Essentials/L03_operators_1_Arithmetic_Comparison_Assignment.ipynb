# Python Operators part 1
### TYPES:  Arithmetic,  Comparison, Assignment

### 1.  Arithmetic Operator +, - , *, /, //, %, **

```python
# Addition +
x = 7
y = 5
res = x + y

print(res) 
```

```python
# Subtraction, multiplication
x = 6
y = 5

print(x - y)
print(x * y)
```

```python
#Now lets perform Normal Division / , Modulo Division //, Remainder Division % 
x = 17
y = 5

print(x / y)  # Normal Division (/)
print(x // y) # Modulo Division (//) --> Floor Division
print(x % y)  # Remainder Division (%)
```

```python
# My favorite: Exponentiation (**)

# STEP1: small power
x = 3
y = 4

print(x ** y)
```

```python
# STEP2: Increase the power
x = 3
y = 40

print(x ** y)
```

```python
# STEP3: Lets crank it up
x = 3
y = 4000

print(x ** y)
```

### 2. Comparison Operators <,<=,>,>=,==,!=
- Comparing data values. Data can be int, string, float ,etc
- Always returns **boolean variable**: True or False

```python
# Lets print the boolean variable: True / False

print(True) 
print(False)
```

```python
# >
x = 16
y = 15
res = x > y # this is a boolean type

print(res) 
# print(type(res))
```

```python
# or I can print the value directly
x = 16
y = 15

print(x > y) 
```

```python
# <
x = 16
y = 15

print(x < y) 
```

```python
x = 13
y = 15

print(x < y) 
```

```python
#  greater than or equal to  >=
x = 15
y = 15

print(x >= y) 
```

```python
# less than or equal to  <=

x = 13
y = 15

print(x <= y) 
```

```python
# is equal to == , not equal to !=

# example1
x = 15
y = 15

print(x == y) 
print(x != y) 
```

```python
# example2
x = 17
y = 15

print(x == y) 
print(x != y) 
```

```python
# We can also compare strings
x = "a"
y = "d"

print(x > y) #string comparision is done lexicographically: a < b < c < d....

##############
# SIDE NOTE: Python uses ord() value of characters 
print(ord(x)) # 97
print(ord(y)) # 100
```

```python
x = "abc"
y = "abd"

print(x > y) # lexicograph
##############################
x = "abczz"
y = "abd"

print(x > y) # lexicograph
```

```python
a = "bobit"
b = "boss"
print(a > b) # lexicograph
```

### 3. Assignment Operator =, +=, -=, *=, %=

```python
# assign to variable a the value of 5
a = 5

print(a)
```

```python
# c could be salary, raise a
c = 12
a = 3

c = c + a # Increment the value of c by a: overwrites the previous value of c
print(c)

# c += a  # c = c + a
# print (c)
```

```python
# c could be salary, deduction a
c = 12
a = 3

c = c - a # Decrement the value of c by a: overwrites the previous value of c
print(c)

# c -= a  # c = c - a
# print (c)
```

```python
c = 12
a = .3

c *= a # c = c * a
print(c)
```

```python
# Remainder
c = 10
a = 7

c %= a # c = c % a 
print(c)
```

```python
c = 10
a = 4

c /= a # c = c / a
print(c)
```

```python
# SKIP: OPTIONAL: compares list
a = [1,2,3]
b = [1,2,3,4]
print(b<a)

####################
# SKIP
x = chr(12)
y = chr(13)

b = x < y

print(b)
```

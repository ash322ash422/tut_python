# ASSIGNMENT: CREATE A SIMPLE STUDENT CLASS

OBJECTIVE:
Create a Student class that stores student details and performs following basic operations using class concepts:
- add_mark
- calculate_average
- display_student

  
CONCEPTS USED:
- Class as a blueprint
- __init__ constructor
- Instance variables
- Instance methods


## Step 1: Create the Student class

```python

class Student:
    # Constructor to initialize student details
    def __init__(self, student_id, name, age):
        # Instance variables
        self.student_id = student_id
        self.name = name
        self.age = age
        self.marks = []
        print("Instance variable created: ", self.student_id, self.name)

    # Method to add marks
    def add_mark(self, mark):
        # Append mark to the marks list
        self.marks.append(mark)

    # Method to calculate average marks
    def calculate_average(self):
        # Check if marks list is empty
        if len(self.marks) == 0:
            return 0
        return sum(self.marks) / len(self.marks)

    # Method to display student details
    def display_student(self):
        print("Student ID:", self.student_id)
        print("Name:", self.name)
        print("Age:", self.age)
        print("Marks:", self.marks)
        print("Average Marks:", self.calculate_average())
```

## Step 2: Create Student objects

```python

student1 = Student(101, "Ash", 21)
student2 = Student(102, "Emma", 22)
```

## Step 3: Add marks to students

```python

student1.add_mark(85)
student1.add_mark(90)
student1.add_mark(88)

student2.add_mark(78)
student2.add_mark(82)
student2.add_mark(80)
```

## Step 4: Display student details


```python

print("Student 1 Details:")
student1.display_student()

print("\nStudent 2 Details:")
student2.display_student()
```

## ASSIGNMENT EXTENSION TASKS (FOR STUDENTS)

1. Add a method to find highest mark
2. Add a method to check pass/fail (average >= 40)
3. Add a method to update student age
4. Create multiple Student objects using user input
5. Store Student objects in a list


## SOLUTION: STUDENT CLASS EXTENSION TASKS

```python


class Student:
    # Constructor to initialize student details
    def __init__(self, student_id, name, age):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.marks = []

    # Method to add marks
    def add_mark(self, mark):
        self.marks.append(mark)

    # Method to calculate average marks
    def calculate_average(self):
        if len(self.marks) == 0:
            return 0
        return sum(self.marks) / len(self.marks)

    # Extension 1: Find highest mark
    def highest_mark(self):
        if len(self.marks) == 0:
            return None
        return max(self.marks)

    # Extension 2: Check pass or fail
    def check_pass_fail(self):
        if self.calculate_average() >= 40:
            return "Pass"
        else:
            return "Fail"

    # Extension 3: Update student age
    def update_age(self, new_age):
        self.age = new_age

    # Display student details
    def display_student(self):
        print("Student ID:", self.student_id)
        print("Name:", self.name)
        print("Age:", self.age)
        print("Marks:", self.marks)
        print("Average:", self.calculate_average())
        print("Highest Mark:", self.highest_mark())
        print("Result:", self.check_pass_fail())
        print("---------------------------------")


# =========================================================
# Extension 4 & 5: Create multiple students using user input
# and store them in a list
# =========================================================

students = []

number_of_students = int(input("Enter number of students: "))

for i in range(number_of_students):
    print("\nEnter details for student", i + 1)

    student_id = int(input("Student ID: "))
    name = input("Student Name: ")
    age = int(input("Student Age: "))

    # Create Student object
    student = Student(student_id, name, age)

    # Add marks for the student
    marks_count = int(input("How many marks to enter?: "))

    for j in range(marks_count):
        mark = float(input("Enter mark " + str(j + 1) + ": "))
        student.add_mark(mark)

    # Add student object to list
    students.append(student)


# =========================================================
# Display all students
# =========================================================

print("\nSTUDENT DETAILS SUMMARY\n")

for student in students:
    student.display_student()


# =========================================================
# Demonstrating age update feature
# =========================================================

if len(students) > 0:
    print("\nUpdating age of first student...")
    students[0].update_age(students[0].age + 1)
    students[0].display_student()
```

# PYTHON LISTS AND TUPLES

This code demonstrates:

- Creating lists

- Indexing and slicing

- append() and remove()

- Tuples and immutability


```python
# CREATING LISTS
# Lists are ordered and mutable (can be changed)

fruits = ["mango", "banana", "cherry", "kiwi"]

print("Original fruits list:")
print(fruits)
```

```python
# INDEXING: Indexing starts from 0
fruits = ["mango", "banana", "cherry", "kiwi"]

print("\nINDEXING EXAMPLES")
print(f"First fruit: {fruits[0]}")
print(f"Second fruit: {fruits[1]}")
print(f"Last fruit: {fruits[-1]}")
```

```python
# SLICING: Slicing gets a portion of the list

fruits = ["mango", "banana", "cherry", "kiwi"]

print("\nSLICING EXAMPLES")
print(f"First two fruits: {fruits[0:2]}")
print(f"Fruits from index 1 to end: {fruits[1:]}")
print(f"All fruits: {fruits[:]}")
```

```python
# append() METHOD: append() adds an item to the end of the list

fruits = ["mango", "banana", "cherry", "kiwi"]

print("\nAPPEND EXAMPLE")
fruits.append("orange")
print(f"After appending orange: {fruits}")
```

```python
# remove() METHOD: remove() deletes a specific item from the list

fruits = ["mango", "banana", "cherry", "kiwi"]

print("\nREMOVE EXAMPLE")
fruits.remove("banana")
print(f"After removing banana: {fruits}")
```

```python
# MODIFYING LIST VALUES
# Lists allow modification using index

fruits = ["mango", "banana", "cherry", "kiwi"]

fruits[0] = "grapes"
print("\nAfter modifying first element:")
print(fruits)
```

```python
# TUPLES
# Tuples are ordered but immutable (cannot be changed)

colors = ("red", "green", "blue", "yellow")

print("\nTUPLE EXAMPLES")
print(colors)
print(f"First color: {colors[0]}")
print(f"Slice of tuple: {colors[1:]}")
```

```python
# IMMUTABILITY OF TUPLES

# The following line would cause an error
# colors = ("red", "green", "blue", "yellow")
# colors[2] = "orange"

print("\nTuples cannot be modified once created.")
print("Lists can be modified, tuples cannot.")
```

### REAL-WORLD COMPARISON
# Use lists when data needs to change
# Use tuples when data should remain fixed

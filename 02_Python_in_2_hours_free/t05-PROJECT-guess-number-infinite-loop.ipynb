# PROJECT: GUESS THE NUMBER GAME

This program generates a secret number and keeps asking the user to guess it until the correct number is guessed.

Concepts covered:
- while True (infinite loop)
- input()
- Typecasting
- if, elif, else
- break
- Comparison operators



```python
# IMPORT REQUIRED MODULE
# random module is used to generate a random number

import random

# GAME SETUP: Generate a random number between 1 and 15
secret_number = random.randint(1, 15)

# Keep track of number of attempts
attempts = 0

print("Welcome to the Guess the Number Game!")
print("I have selected a number between 1 and 15.")


# INFINITE LOOP: The loop runs until the correct guess is made
while True:
    # Take user input
    guess_input = input("Enter your guess: ")
    
    # Convert input to integer
    guess = int(guess_input)
    
    # Increase attempt count
    attempts = attempts + 1
    
    # CHECK THE GUESS
    if guess == secret_number:
        print("Congratulations! You guessed the correct number.")
        print(f"Number of attempts: {attempts}")
        break   # Exit the loop when correct
    
    elif guess < secret_number:
        print("Too low! Try again.")
    
    else:
        print("Too high! Try again.")


# GAME OVER MESSAGE
print("Game Over. Thanks for playing!")
```

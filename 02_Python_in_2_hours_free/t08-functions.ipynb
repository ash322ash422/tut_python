# FUNCTIONS IN PYTHON
Topics Covered:
- def keyword
- Parameters vs Arguments
- return vs print
- Variable Scope (Global vs Local)


```python
# def KEYWORD

# def is used to define (create) a function
def greet():
    print("Hello! Welcome to Python functions.")
    print("You will like this programming language")

# Calling the function
greet()
```

```python
# PARAMETERS vs ARGUMENTS

# Parameters are variables listed in the function definition
def greet_user(name, age):
    # name and age are PARAMETERS
    print("Name:", name)
    print("Age:", age)

# Arguments are actual values passed to the function
greet_user("Ash", 25)   # "Ash" and 25 are ARGUMENTS
```

# return vs print

```python
# Function using print (shows output but does not return data)
def add_print(a, b):
    print("Sum using print:", a + b)
    print("Function executed sucessfully")

result_print = add_print(10, 20)
print("Value returned from add_print:", result_print)
```

```python
# Function using return (sends data back to caller)
def add_return(a, b):
    return a + b

result_return = add_return(10, 20)
print("Sum using return:", result_return)
```

# GLOBAL vs LOCAL SCOPE

```python
# Global variable (accessible everywhere)
discount = 10

def calculate_price(price):
    # price is a local variable
    final_price = price - discount
    return final_price

print("Final price:", calculate_price(100))
print("Discount (Global):", discount)
```

```python
# LOCAL VARIABLE EXAMPLE

def local_example():
    message = "This is a local variable"
    print(message)

local_example()

# The following line will cause an error if uncommented
# print(message)
```

```python
# MODIFYING GLOBAL VARIABLE

count = 0

def increment():
    global count # if you want to modify global var, then use keyword global
    count = count + 1

increment()
increment()
print("Global count after increment:", count)
```

# LOADING A CSV FILE USING PANDAS

```python
# pip install pandas
# pip install matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load a CSV file into a DataFrame
# Make sure a CSV file exists in the same directory
# Example file name: data_iris.csv

df = pd.read_csv("data_iris.csv")

# Display first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

## BASIC DATA INSPECTION

```python
# Check shape of the DataFrame
print("\nShape of dataset (rows, columns):")
print(df.shape)

# Check column names
print("\nColumn names:")
print(df.columns)
```

# df.describe()


```python
# describe() gives statistical summary of numerical columns
# Includes count, mean, std, min, max, and quartiles

print("\nStatistical summary using describe():")
print(df.describe())
```

## PLOTTING A SIMPLE GRAPH

```python
# Simple line plot using one numerical column
# Replace 'column_name' with an actual column from your CSV

numeric_column = df.select_dtypes(include="number").columns[0]

print("\nPlotting column:", numeric_column)

# Plot the data
df[numeric_column].plot()

# Add labels and title
plt.xlabel("Index")
plt.ylabel(numeric_column)
plt.title("Simple Line Plot using Pandas")

# Display the plot
plt.show()
```

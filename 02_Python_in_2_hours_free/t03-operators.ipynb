# PYTHON OPERATORS

This code demonstrates:
- Arithmetic operators
- Comparison operators
- Logical operators


```python
# ARITHMETIC OPERATORS
# +  addition
# -  subtraction
# *  multiplication
# /  division
# % remainder

a = 20
b = 7

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b
remainder = a % b  

print("ARITHMETIC OPERATIONS")
print(f"a + b = {addition}")
print(f"a - b = {subtraction}")
print(f"a * b = {multiplication}")
print(f"a / b = {division}")
print(f"a % b = {remainder}")
```

```python
# COMPARISON OPERATORS
# ==  equal to
# >   greater than
# <   less than

a = 20
b = 7

print("\nCOMPARISON OPERATIONS")

print(f"a == b : {a == b}")
print(f"a > b  : {a > b}")
print(f"a < b  : {a < b}")
```

```python
# LOGICAL OPERATORS
# and -> True if both conditions are True
# or  -> True if at least one condition is True

age = 25
has_id = True # True , False

print("\nLOGICAL OPERATIONS")

# Using AND operator
is_allowed_and = (age >= 18) and has_id
# is_allowed_and = True      and True

print(f"Allowed with AND condition: {is_allowed_and}")


# Using OR operator
is_allowed_or = (age >= 18) or has_id
print(f"Allowed with OR condition: {is_allowed_or}")
```

```python
# REAL-WORLD STYLE EXAMPLE

# A person can enter if:
# - Age is greater than 18 AND has an ID

age = 25
has_id = True

can_enter = (age > 18) and has_id

print("\nFINAL DECISION")

print(f"Can enter: {can_enter}")
```

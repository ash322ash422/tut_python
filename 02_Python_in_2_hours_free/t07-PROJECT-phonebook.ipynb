# PROJECT TITLE:
Simple Phonebook Application

### PROJECT DESCRIPTION:
Build a small phonebook system that stores contact names
with their phone numbers and allows searching for a contact
by name.

CONCEPTS TO BE USED:
- Dictionary (key-value pairs)
- User input
- Searching data


```python
# Step 1: Create an empty phonebook
phonebook = {}

print("Welcome to the Phonebook Application")


# Step 2: Add contacts to the phonebook
# The user will enter name and phone number
# Type 'stop' as name to finish adding contacts

while True:
    name = input("Enter contact name (or type 'stop' or 'q' to quit): ")

    if name.lower() == "stop" or name.lower() == 'q':
        break

    phone_number = input("Enter phone number: ")

    # Store name and phone number as key-value pair
    phonebook[name] = phone_number

    print("Contact saved successfully\n")


# Step 3: Display all saved contacts
print("All contacts in phonebook:")
print(phonebook)


# Step 4: Search for a contact by name
# The user can look up a phone number using the contact name

search_name = input("\nEnter a name to search:")

# Check if the name exists in the phonebook
if search_name in phonebook:
    print("Phone number for", search_name, "is:", phonebook[search_name])
else:
    print("Contact not found")


# Step 5: Safe search using get()
# This avoids errors if the contact does not exist

result = phonebook.get(search_name)
print("Search result using get():", result)
```

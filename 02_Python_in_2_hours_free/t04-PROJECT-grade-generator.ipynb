# PROJECT: GRADE GENERATOR

This program takes a student's score and generates a grade (A / B / C / Fail)
```
GRADE LOGIC: Grade rules:
90 - 100 -> A
75 - 89  -> B
60 - 74  -> C
Below 60 -> Fail
```

Concepts covered:
- input()
- Typecasting
- if, elif, else
- Comparison operators
- f-strings

````python
# USER INPUT
# input() always returns a string

student_name = input("Enter student name: ")
score_input = input("Enter score (0 to 100): ")

# TYPECASTING: Convert score from string to integer
score = int(score_input)

# GRADE LOGIC: Grade rules:
# 90 - 100 -> A
# 75 - 89  -> B
# 60 - 74  -> C
# Below 60 -> Fail

if score >= 90 and score <= 100:
    grade = "A"
elif score >= 75 and score < 90:
    grade = "B"
elif score >= 60 and score < 75:
    grade = "C"
else:
    grade = "Fail"

# OUTPUT RESULT
print("\n----- GRADE REPORT -----")
print(f"Student Name: {student_name}")
print(f"Score: {score}")
print(f"Grade: {grade}")
````

````python
# Simple real-world message

if grade == "A":
    print("Excellent performance!")
elif grade == "B":
    print("Good job, keep improving.")
elif grade == "C":
    print("You passed, but aim higher.")
else:
    print("You need to work harder.")
````

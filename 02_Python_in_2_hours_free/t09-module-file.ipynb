# IMPORTING BUILT-IN MODULES
Topics:
- import math
- import random

```python
# Import math module for mathematical operations
import math

# Import random module for random number generation
import random
```

```python
# USING math MODULE

number = 16

# Square root of a number
print("Square root of", number, "is:", math.sqrt(number))

# Power calculation
print("2 raised to power 5:", math.pow(2, 5))

# Value of pi
print("Value of pi:", math.pi)
```

```python
# USING random MODULE

# Generate a random integer between 1 and 10
random_number = random.randint(1, 10)
print("Random number between 1 and 10:", random_number)

# Random choice from a list
colors = ["red", "green", "blue", "yellow"]
random_color = random.choice(colors)
print("Randomly selected color:", random_color)
```

# TRY / EXCEPT (ERROR HANDLING)

```python
#  Example 1: Handling division by zero
# 1) Without error handling: problem user enters 0 (ZeroDivisionError) or maybe a string (ValueError)

value = int(input("\nEnter a number to divide 100 by: "))
result = 100 / value
print("Result:", result)
```

```python
# 2) 
try:
    value = int(input("\nEnter a number to divide 100 by: "))
    result = 100 / value
    print("Result:", result)
except ZeroDivisionError:
    print("Errrror: Division by zero is not allowed")
except ValueError:
    print("Errrror: Please enter a valid number")
else:
    print("Calculation successful")
finally:
    print("Try-except block finished")
```

# FILE I/O (READ AND WRITE)

```python
# Step 1: Writing data to a file
# 'w' mode creates a file if it does not exist
file = open("sample_data.txt", "w")

file.write("This is a sample file.\n")
file.write("Python File I/O example.\n")
file.write("Learning open, read, and write.\n")
file.write("I am writing to file\n")

file.close()
print("\nData written to file successfully")
```

```python
# Step 2: Reading data from a file
# 'r' mode is used to read a file
file = open("sample_data.txt", "r")

content = file.read()
print("\nFile content:")
print(content)

file.close()
```

## FILE I/O USING **WITH** STATEMENT (BEST PRACTICE)

```python
# The 'with' statement automatically closes the file
# 1) appending data to the file

with open("sample_data.txt", "a") as file:
    file.write("Appending new data to the file.\n")

print("New data appended successfully")
```

```python
# Read again to verify appended content
with open("sample_data.txt", "r") as file:
    print("\nUpdated file content:")
    content = file.read() 
    print(content)
```

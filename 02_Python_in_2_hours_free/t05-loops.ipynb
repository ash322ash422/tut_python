# PYTHON LOOPS

This code demonstrates:

- for loop with range()
- while loop
- break and continue

```python
# FOR LOOP WITH range()
# range(start, stop) generates numbers
# stop value is not included
# range(3, 10) -> 3, 4, 5, 6, 7, 8, 9

print("FOR LOOP EXAMPLE")

for i in range(10, 12): # is called iterator
    print(f"Iteration number: {i}")

print("THANK YOU. Please come again.")
```

```python
# WHILE LOOP
# while loop runs as long as condition is True

print("\nWHILE LOOP EXAMPLE")

count = 1

while count <= 5:
    print(f"Count is: {count}")
    count = count + 1   # Update condition to avoid infinite loop

print("Thank you. for visiting while loop")
```

```python
# BREAK STATEMENT
# break stops the loop completely

print("\nBREAK STATEMENT EXAMPLE")

for num in range(1, 10):
    if num == 5:
        print("Breaking the loop at 5")
        break
    print(f"Number: {num}")
```

```python
# CONTINUE STATEMENT
# continue skips the current iteration

print("\nCONTINUE STATEMENT EXAMPLE")

for num in range(1, 8): # i=1, 2, 3, 4, 5, 6, 7
    if num == 4:
        print("Skipping number 4")
        continue
    print(f"Number: {num}")
```

```python
# REAL-WORLD STYLE EXAMPLE
# Find first number divisible by 7

print("\nREAL-WORLD LOOP EXAMPLE")

for num in range(13, 50):
    if num % 7 == 0:
        print(f"First number divisible by 7 is: {num}")
        break
```

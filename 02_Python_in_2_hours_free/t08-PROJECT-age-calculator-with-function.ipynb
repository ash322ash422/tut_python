# PROJECT: REFACTORING AGE CALCULATOR INTO A FUNCTION

### PROJECT DESCRIPTION: 
Convert the age calculation logic into a reusable function
so it can be used multiple times in a program.



## Original age calculator without function

```python
# Step 1: Ask user for year of birth
birth_year = int(input("Enter your year of birth: "))

# Step 2: Define the current year
current_year = 2026

# Step 3: Calculate age: example 2026 - 2000 = 26 years old
age = current_year - birth_year

# Step 4: Display the age
print("Your age is:", age)
```

## New age calculator with function

```python
# Step 1: Create a function using def keyword

current_year = 2026

def calculate_age(birth_year, current_year):
    # birth_year and current_year are parameters
    # This function returns the calculated age
    return current_year - birth_year


# Step 2: Call the function with arguments
user_birth_year = int(input("\nEnter your year of birth again: "))

# Call the reusable function
user_age = calculate_age(user_birth_year, current_year)

# Step 3: Display the returned value
print("Your age using function is:", user_age)
```

# FUNCTION REUSABILITY EXAMPLE

```python
# Calculate age for another person

friend_birth_year = 1995
friend_age = calculate_age(friend_birth_year, current_year)
print("Friend's age is:", friend_age)
```

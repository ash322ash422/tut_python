```python
# ================================

# PYTHON BASICS DEMO CODE


# VARIABLES = CONTAINERS

# -------------------------------

# A variable stores (or contains) a value in memory

name = "Ash"          # String variable
age = 30              # Integer variable
height = 5.9          # Float variable
is_trainer = True     # Boolean variable
```

```python

# -------------------------------

# DATA TYPES: int, float, string, bool

# -------------------------------

# int  -> whole numbers

marks = 85

# float -> decimal numbers

percentage = 85.5

# string -> text data

course = "Python Programming"

# bool -> True or False

is_passed = True
```

```python
# -------------------------------

# PRINTING VALUES AND TYPES

# -------------------------------

print(name)
print(age)
print(height)
print(is_trainer)

# type() shows the data type of a variable

# print(type(marks))
# print(type(percentage))
# print(type(course))
# print(type(is_passed))
```

```python
# -------------------------------

# F-STRINGS (Formatted Strings)

# -------------------------------

# f-strings allow us to insert variables directly inside strings

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Height: {height}")
print(f"Trainer Status: {is_trainer}")

# You can also do calculations inside f-strings

print(f"Age after 5 years: {age + 5}")

```

```python
# -------------------------------

# input() FUNCTION

# -------------------------------

# input() always returns data as a STRING

user_name = input("Enter your name: ")
user_age  = input("Enter your age: ")

print(f"User Name: {user_name}")
print(f"User Age (string): {user_age}")
print(type(user_age))   # This will be <class 'str'>
```

```python
# -------------------------------

# TYPECASTING

# -------------------------------

# Converting one data type to another

user_age_int = int(user_age)    # Convert string to integer
user_height = float(input("Enter your height: "))  # Direct typecasting

# Boolean typecasting example

is_adult = user_age_int >= 18

print(user_age_int, user_height )
```

```python
# -------------------------------

# USING TYPECASTED VALUES

# -------------------------------

print(f"Age next year: {user_age_int + 1}")
print(f"Height entered: {user_height}")
print(f"Is adult: {is_adult}")
```

```python
# -------------------------------

# SUMMARY OUTPUT

# -------------------------------

print("----- SUMMARY -----")
print(f"Name: {user_name}")
print(f"Age: {user_age_int}")
print(f"Height: {user_height}")
print(f"Adult Status: {is_adult}")

# End of Python basics demo code
```

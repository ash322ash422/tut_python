# PYTHON CONDITIONAL STATEMENTS DEMO

This code demonstrates:

- if, elif, else

- Indentation (physical lines matter in Python)

- Nested conditions


```python
# USER INPUT
# input() always returns a string

name = input("Enter your name: ")
age = int(input("Enter your age: "))
marks = int(input("Enter your marks (0 to 100): "))

# IF - ELIF - ELSE EXAMPLE
# Python checks conditions from top to bottom
# Only one block will execute

print("\nRESULT ANALYSIS")

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
elif marks >= 40:
    grade = "D"
else:
    grade = "F"

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Grade: {grade}")
```

```python
# NESTED IF CONDITIONS
# if inside another if block


print("\nELIGIBILITY CHECK")

if age >= 18:
    # This block runs only if age >= 18
    print("You are 18 or older.")
    
    if grade != "F":
        # This block runs only if age >= 18 AND grade is not F
        print("You are eligible for the course.")
    else:
        print("You are not eligible due to low grade.")

else:
    # This block runs if age < 18
    print("You are under 18.")
    print("You are not eligible for the course.")
```

```python
# INDENTATION EXPLANATION
# Indentation defines code blocks in Python
# Wrong indentation will cause errors or wrong output

age = 10

print("\nINDENTATION DEMO")

if age > 20:
    print("This line is inside the if block.")
    print("Both lines run because indentation is correct.")
print("This line is outside the if block.")
```

```python
# Copy and paste the code from file 4 step1

# # STEP1: lets iterate over the whole file and print all the employees
# Each new employee entry starts at i = 0, 14, 28, ..

file_name = "swipe_schedule.txt"
file = open(file_name, "r")
lines = file.readlines()  # Reads all lines into a list
file.close()

number_of_employees = int(len(lines)/14) 

for i in range(0, number_of_employees): # i = 0, 1, 2, 3
    l = lines[i*14: i*14 + 14]  # <- lines[0:14], lines[14:28], lines[28:42], lines[42:56]
    print(l)
    print("++++++++++++++++++++++++++++++++++++++++++++\n")

```

```python
# step2: Lets print data of employee based on employee id = c0096

search_employee_id = 'c0095'

file_name = "swipe_schedule.txt"
file = open(file_name, "r")
lines = file.readlines()  # Reads all lines into a list
file.close()

number_of_employees = int(len(lines)/14) 

for i in range(0, number_of_employees): # i = 0, 1, 2, 3
    l = lines[i*14: i*14 + 14]  # <- lines[0:14], lines[14:28], lines[28:42], lines[42:56]
    employee_id =  l[6].split(":")[1].strip()
    
    if search_employee_id == employee_id:
        print(l)
```

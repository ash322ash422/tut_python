```python
# Step2: Now extract the values and store them in list of dictionary:
# lets create a dictionary to store all employees data. Then store the dict into a list

file_name = "swipe_schedule.txt"
file = open(file_name, "r")

lines = file.readlines()  # Reads all lines into a list
number_of_employees = int(len(lines)/14) 

records = []
record  = {}
for i in range(0, number_of_employees): # i = 0, 1, 2, 3
    l = lines[i*14: i*14 + 14]  # <- lines[0:14], lines[14:28], lines[28:42], lines[42:56]
    print("l=", l)
    record['date_today']  = l[1].split(":")[1].strip()
    record['first_name']  = l[3].split(":")[1].strip()
    record['middle_name'] = l[4].split(":")[1].strip()
     
    records.append(record)
    print("+++++++++++++++++++++++++++++++++++\n")
    
print("\n\nrecords:", records)
file.close()
```

```python
# (OPTIONAL: ALternative to above) STEP3: lets create a dictionary to store all employees data. Then store the dict into a list

file_name = "swipe_schedule.txt"
file = open(file_name, "r")

lines = file.readlines()  # Reads all lines into a list

records = []
record = {}

for line in lines:
    line = line.strip()
    if not line:
        continue  # skip empty lines

    # New record starts
    if line.startswith("--------------- record"):
        if record:  # save previous record
            records.append(record)
            record = {}
        continue

    if line.startswith("date today:"):
        record["date_today"] = line.split(":", 1)[1].strip()
    elif line.startswith("first name:"):
        record["first_name"] = line.split(":", 1)[1].strip()
    elif line.startswith("middle name:"):
        record["middle_name"] = line.split(":", 1)[1].strip()
    elif line.startswith("last name:"):
        record["last_name"] = line.split(":", 1)[1].strip()
    elif line.startswith("employee id:"):
        record["employee_id"] = line.split(":", 1)[1].strip()
    elif line.startswith("date of join:"):
        record["date_of_join"] = line.split(":", 1)[1].strip()
    elif line.startswith("card swipes:"):
        swipes = line.split(":", 1)[1].strip()
        record["card_swipes"] = [s.strip() for s in swipes.split(",") if s.strip()]
    elif line.startswith("group:"):
        record["group"] = line.split(":", 1)[1].strip()
    elif line.startswith("comments:"):
        record["comments"] = line.split(":", 1)[1].strip()

# Add last record
if record:
    records.append(record)

print(records)

file.close()
```

```python
# STEP1: lets find the index of date of join for employees from the list "lines"

file_name = "swipe_schedule.txt"
file = open(file_name, "r")
lines = file.readlines()  # Reads all lines into a list
file.close()

print(lines)
print("length:",len(lines)) # this gives me the total length of list

print(lines[7])  # 7 =  7 + 0*14
print(lines[21]) # 21 = 7 + 1*14
print(lines[35]) # 35 = 7 + 2*14
print(lines[49]) # 49 = 7 + 3*14 
# print(lines[63]) # out of range error
```

```python
# STEP2: lets create a for loop that prints date of join and it automatically calculates
# how many lines to print

num_of_employees = int(len(lines)/14)

for i in range(0, num_of_employees): # i = 0, 1, 2, 3, 4
    print(lines[7 + i*14])  # 7 =  7 + i*14
```

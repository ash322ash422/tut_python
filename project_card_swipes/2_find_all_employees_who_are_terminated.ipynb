```python
# STEP1: lets iterate over the whole file and print all the employees
# i = 0, 14, 28, ..

file_name = "swipe_schedule.txt"
file = open(file_name, "r")

lines = file.readlines()  # Reads all lines into a list
number_of_employees = int(len(lines)/14) 
print(number_of_employees)

for i in range(0, number_of_employees): # i = 0, 1, 2, 3
    l = lines[i*14: i*14 + 14]  # <- lines[0:14], lines[14:28], lines[28:42], lines[42:56]
    rec_str = ""
    for i in l:
        rec_str = rec_str + i
        
    print(rec_str)
    print("++++++++++++++++++++++++++++++++++++++++++++")
file.close()
```

```python
# STEP2: lets iterate over the whole file and print only those employees who are terminated

file = open(file_name, "r")

lines = file.readlines()  # Reads all lines into a list
number_of_employees = int(len(lines)/14) 

for i in range(0, number_of_employees): # i = 0, 1, 2, 3
    l = lines[i*14: i*14 + 14]  # <- lines[0:14], lines[14:28], lines[28:42], lines[42:56]

    comments = l[-1] # this would give me the last entry which is comments
    if 'terminated' in comments.lower():
        rec_str = ""
        for i in l:
            rec_str = rec_str + i
            
        print(rec_str)
    
    # print("++++++++++++++++++++++++++++++++++++++++++++")
file.close()
```

```python
file_name = "swipe_schedule.txt"
```

```python
# step1: Read All Lines into a List and print the entire list

#step1
file = open(file_name, "r")

lines = file.readlines()  # Reads all lines into a list
print(f"lines: \n {lines}")
print(f"type: {type(lines)}")
print(f"number of lines: {len(lines)}")
print(f"number of employees: {len(lines)/14}")

file.close()
```

### SOLUTION1: By looking at the structure of file, I notice the each record is 14 lines long.
- Records number 1: line 1 - 14
- Records number 2: line 15 - 28
- Records number 3: line 29 - 42


```python
# step1: Print the records for first 3 employees
# So I can use slicing: [start_idx: stop_idx: step_size]

l = lines[0:14]
print(l)

l = lines[14:28]
print(l)

l = lines[28:42]
print(l)
```

```python
# step2: concatenate the strings in each list. Start with first list
# Start by concatenating the first list or first employee

l = lines[0:14]
print(f"l: {l}")

rec_str = ""
for i in l:
    rec_str = rec_str + i
    
print(rec_str)
```

```python
# step3: Now do this for all 3 lists

l = lines[0:14]
rec_str = ""
for i in l:
    rec_str = rec_str + i
    
print(rec_str)
print("++++++++++++++++++++++++++++++++++++++++++++")

l = lines[14:28]
rec_str = ""
for i in l:
    rec_str = rec_str + i
    
print(rec_str)
print("++++++++++++++++++++++++++++++++++++++++++++")

l = lines[28:42]
rec_str = ""
for i in l:
    rec_str = rec_str + i
    
print(rec_str)
```

```python
#step4: generalize . Lets put above code in 
# i = 0, 14, 28 

for i in [0, 14, 28]:
    l = lines[i:i+14]  # <- This is what is being used to change l
    rec_str = ""
    for i in l:
        rec_str = rec_str + i
        
    print(rec_str)
    # print("++++++++++++++++++++++++++++++++++++++++++++")
```

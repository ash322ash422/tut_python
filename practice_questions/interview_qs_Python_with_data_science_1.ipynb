### 2. How do you find missing data in pandas?

df.isnull().sum()


### 2. How do you handle missing data in pandas?

Answer:

Drop missing values: df.dropna()

Fill missing values: df.fillna(value)

Forward fill: df.fillna(method="ffill")

Backward fill: df.fillna(method="bfill")

Replace with mean/median/mode: df["col"].fillna(df["col"].mean())


### 3. How do you select rows and columns in pandas?

Answer:

By label: df.loc[2, "col_name"]

By index: df.iloc[2, 0]

Multiple rows/cols: df.loc[2:5, ["col1", "col2"]]


### 4. What is the difference between .loc[] and .iloc[] in pandas?

Answer:

.loc[] → label-based indexing (row/column names).

.iloc[] → position-based indexing (integer indices).


### 5. How do you find correlation between columns in pandas?

Answer:

correlation coeff ranges from -1 to + 1

Use .corr() method:

Pearson → linear correlation.

Spearman → rank correlation.
```
df.corr(method="pearson")   # default
df.corr(method="spearman")
```

### 6. How do you find and remove duplicate rows in a DataFrame?

Answer:
df.duplicates()

df.drop_duplicates()

Keep first/last: df.drop_duplicates(keep="first")

Subset columns: df.drop_duplicates(subset=["col1", "col2"])



### 7. How do you group data in pandas?

### 10. How would you check if a dataset is normally distributed?

Answer:

Histogram / KDE plot (visual check).

Statistical tests:
```
from scipy.stats import shapiro, normaltest

shapiro(df["col"])   # Shapiro-Wilk Test
normaltest(df["col"]) # D’Agostino’s K² Test
```

Q-Q Plot:
```
import statsmodels.api as sm
sm.qqplot(df["col"], line="s")
```

Here are 10 multiple-choice questions (MCQs) on lists in Python:

### Question 1
What will be the output of the following code?
```python
l = [1, 2, 3, 4]
print(l[1:3])
```
A) [1, 2]  
B) [2, 3]  
C) [3, 4]  
D) [2, 3, 4]  

**Correct Answer:** B) [2, 3]

---

### Question 2
Which method is used to add an element to the end of a list?

A) add()  
B) append()  
C) insert()  
D) extend()  

**Correct Answer:** B) append()

---

### Question 3
What will be the output of the following code?
```python
l = [1, 2, 3]
l.append([4, 5])
print(l)
```
A) [1, 2, 3, 4, 5]  
B) [1, 2, 3, [4, 5]]  
C) [1, 2, 3, 4, 5, []]  
D) [1, 2, 3, 4, 5]  

**Correct Answer:** B) [1, 2, 3, [4, 5]]

---

### Question 4
Which of the following methods removes the first occurrence of a specified value from a list?

A) delete()  
B) remove()  
C) pop()  
D) discard()  

**Correct Answer:** B) remove()

---

### Question 5
What will be the output of the following code?
```python
l = [1, 2, 3, 4]
print(l[-1])
```
A) 1  
B) 2  
C) 3  
D) 4  

**Correct Answer:** D) 4

### Question 6
Which of the following will correctly sort the list `l = [3, 1, 2]` in ascending order?

A) l.sort()  
B) sorted(l)  
C) l.order()  
D) A and B  

**Correct Answer:** D) A and B

### Question 7
What will be the output of the following code?
```python
l = [1, 2, 3]
l[1] = 4
print(l)
```
A) [1, 2, 3]  
B) [1, 4, 3]  
C) [4, 2, 3]  
D) [1, 2, 4]  

**Correct Answer:** B) [1, 4, 3]

---

### Question 8
Which of the following will create a list containing the numbers 1 to 5?

A) list(range(1, 6))  
B) [1, 2, 3, 4, 5]  
C) Both A and B  
D) None of the above  

**Correct Answer:** C) Both A and B

---

### Question 9
What will be the output of the following code?
```python
l = [1, 2, 3]
l.insert(1, 4)
print(l)
```
A) [1, 4, 2, 3]  
B) [4, 1, 2, 3]  
C) [1, 2, 4, 3]  
D) [1, 2, 3, 4]  

**Correct Answer:** A) [1, 4, 2, 3]

### Question 10
Which of the following statements will create a copy of a list `l`?

A) l.copy()  
B) l[:]  
C) Both A and B  
D) None of the above  

**Correct Answer:** C) Both A and B

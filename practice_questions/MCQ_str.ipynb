Here are 8 multiple-choice questions (MCQs) on strings in Python:

### Question 1
What will be the output of the following code?
```python
s = "Hello, World!"
print(s[7:12])
```
A) Hello  
B) World  
C) , Wor  
D) World  

**Correct Answer:** D) World

---

### Question 2
Which method can be used to convert a string to lowercase in Python?
A) lower()  
B) toLower()  
C) lowercase()  
D) downcase()  

**Correct Answer:** A) lower()

---

### Question 3
What will be the output of the following code?
```python
s = "Python"
print(s * 2)
```
A) PythonPython  
B) Python 2  
C) 2Python  
D) Python + Python  

**Correct Answer:** A) PythonPython

### Question 4
Which of the following methods can be used to find the length of a string in Python?

A) length()  
B) len()  
C) size()  
D) count()  

**Correct Answer:** B) len()

### Question 6
Which of the following will correctly replace "World" with "Python" in the string `s = "Hello, World!"`?
A) s.replace("World", "Python")  
B) s.change("World", "Python")  
C) s.substitute("World", "Python")  
D) s.swap("World", "Python")

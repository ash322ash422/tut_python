# 1. What is Object-Oriented Programming?

# OOP organizes code using:

# 👤 3. Attributes (Variables inside Class)

````python
# Attributes store data related to an object

class Car:
    
    # constructor method
    def __init__(self, name, model_year):
        self.name = name   # instance attribute
        self.model_year = model_year

# Creating objects
car1 = Car("toyota", 2023)
car2 = Car("honda", 2019)

print(car1.name)
print(car1.model_year)

print(car2.name)
print(car2.model_year)
````

````python
# Attributes store data related to an object

class Student:
    
    # constructor method
    def __init__(self, name, age):
        self.name = name   # instance attribute
        self.age = age

# Creating objects
student1 = Student("Ash", 25)
student2 = Student("Riya", 22)

print(student1.name)
print(student1.age)

print(student2.name)
print(student2.age)
````

## Important Concept

self refers to the CURRENT object

- student1.name  → belongs only to student1
- student2.name  → belongs only to student2


# 4. Methods (Functions inside Class)

````python
class Car:
    
    # constructor method
    def __init__(self, name, model_year):
        self.name = name   # instance attribute
        self.model_year = model_year
    
    # method (function inside class)
    def display_info(self):
        print("Car name:", self.name, "; model year:", self.model_year)


c = Car("Mercedez", 1984)

# calling method using object
c.display_info()
````

````python
class Student:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # method (function inside class)
    def introduce(self):
        print("Hi, my name is", self.name)


s = Student("Rahul", 24)

# calling method using object
s.introduce()
````


## Multiple Methods Example


````python
class Calculator:
    
    def add(self, a, b):
        return a + b
    
    def multiply(self, a, b):
        return a * b

calc = Calculator()

print(calc.add(3, 5))
print(calc.multiply(4, 6))
````

````python
from datetime import datetime

class Car:
    
    # constructor method
    def __init__(self, name, model_year):
        self.name = name   # instance attribute
        self.model_year = model_year
    
    # method (function inside class)
    def display_info(self):
        print("Car name:", self.name, "; model year:", self.model_year)

    def display_age(self):
        current_year = datetime.now().year
        age = current_year - self.model_year
        print("Age of car:", age)


c = Car("Mercedez", 2000)

# calling method using object
c.display_info()
c.display_age()
````

````python
from datetime import datetime

class Student:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # method (function inside class)
    def introduce(self):
        print("Hi, my name is", self.name, "and age is", self.age)


    def year_turn_80(self):
        current_year = datetime.now().year
        year_80 = current_year + (80 - self.age)
        print("I will turn 80 in ", year_80)

    
s = Student("Rahul", 24)

# calling method using object
s.introduce()
s.year_turn_80()
````

````python
# 6. Instance Variables vs Class Variables
````

````python
## Instance Variables

class Dog:
    
    def __init__(self, name):
        self.name = name   # instance variable

d1 = Dog("Tommy")
d2 = Dog("Rocky")

print(d1.name)
print(d2.name)


````

## Each object has its OWN copy

## Class Variables

````python
class Dog:
    
    species = "Canine"   # class variable (shared)
    
    def __init__(self, name):
        self.name = name

d1 = Dog("Tommy")
d2 = Dog("Rocky")

print(d1.species)
print(d2.species)
````

## Same value shared across objects

````python
## Accessing Class Variable

print(Dog.species)   # accessed using class name
# print(Dog.name) # ERROR
````

````python
## Changing Class Variable

Dog.species = "Domestic Canine"

print(d1.species)
print(d2.species)
````

# Instance vs Class Variable (Important Demo)

````python
class Counter:
    
    count = 0   # class variable
    
    def __init__(self):
        Counter.count += 1   # increment shared counter
        
c1 = Counter()
c2 = Counter()
c3 = Counter()

print("Total objects created:", Counter.count)
````

 Used for:

* counting objects
* shared configuration
* constants


````python
# Real Example (Student Management)

class Student:
    
    school_name = "ABC School"   # class variable
    
    def __init__(self, name, marks):
        self.name = name          # instance variable
        self.marks = marks
    
    def display(self):
        print(self.name, self.marks, Student.school_name)

s1 = Student("Ash", 90)
s2 = Student("Riya", 85)

s1.display()
s2.display()
````


# OOP Recap

```python
# CLASS      -> Blueprint
# OBJECT     -> Instance of class
# ATTRIBUTE  -> Data (variables)
# METHOD     -> Function inside class
# CONSTRUCTOR-> __init__()
# INSTANCE VARIABLE -> Unique per object
# CLASS VARIABLE    -> Shared by all objects
```

# 1️) Search → Check if pattern exists

- Scans through the entire string and returns the first match of the pattern.
- If a match is found → returns a match object
- If no match → returns None


```python
text = "Loxford is an academy. A new academy"

pattern = r"academy"

match = re.search(pattern, text)
if match:
    print("match:", match)
    print("Found:", match.group())
    print("Start index:", match.start())
    print("End index:", match.end())
else:
    print("Not found")
```

```python
text = """
Hello, my email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit https://example.com for more info.
Call me at +91-9876543210 tomorrow.
#NLP is fun! #AI #MachineLearning
"""

pattern = r"\d+"   # any number

match = re.search(pattern, text)

if match:
    print("Found:", match.group())
else:
    print("Not found")
```

```python
text = "The id number of employee is 432-678910."

pattern = r"\d+-\d+"
match = re.search(pattern, text)

print(match.group())
```

```python
text = """
Date: 2025-06-10
Hello, my email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit https://example.com for more info.
Call me at +91-9876543210 tomorrow.
#NLP is fun! #AI #MachineLearning
"""


pattern = r"(\d{4})-(\d{2})-(\d{2})"

match = re.search(pattern, text)

if match:
    print("Year:", match.group(1))
    print("Month:", match.group(2))
    print("Day:", match.group(3))
```

# 3️) Find All → Extract all matches

```python
text = """
Hello, my email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit https://example.com for more info.
Call me at +91-9876543210 tomorrow.
#NLP is fun! #AI #MachineLearning
"""

# pattern = r"\@" # find all @
# pattern = r"(\w+)" # find all words
# pattern = r"(\d)" # find all digits
pattern = r"(\D)" # all chars including newline, spaces
# pattern = r"."# Matches any character except a newline.
# pattern = r"\b\w+@\w+\.\w+\b"   # emails

results = re.findall(pattern, text)
print("results:", results)
```

```python
# Find word that starts with 'p' and ends with 't'
text = "He made profit of 545 bucks. Is this True ?  "

pattern = r"\bp\w+t\b"

results = re.findall(pattern, text)
print("results:", results)
```

```python
text = "The email of alice is alice-b@google.com or al@hotmail.com"

pattern = r'[\w\.-]+@[\w\.-]+'

results = re.findall(pattern, text)
print("results:", results)
```

# 4️) Find Iteratively → With positions

```python
text = """
Hello, my email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit https://example.com for more info.
Call me at +91-9876543210 tomorrow.
#NLP is fun! #AI #MachineLearning
"""

pattern = r"#\w+"   # hashtags

for match in re.finditer(pattern, text):
    print(match.group(), "at position", match.start(), "-", match.end())
```

# 5️) Split → Tokenization-like behavior

```python
text = """
Hello, my email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit https://example.com for more info.
Call me at +91-9876543210 tomorrow.
#NLP is fun! #AI #MachineLearning
"""

pattern = r"\s+"   # split on whitespace

tokens = re.split(pattern, text)
print(tokens[:10])  # show first 10 tokens
```

# 6️) Substitution → Text Cleaning (MOST IMPORTANT)

```python
text = "I love cats. Cats are cute."

pattern = r"cats"

new_text = re.sub(pattern, "dogs", text)
new_text = re.sub(pattern, "dogs", text, flags=re.IGNORECASE)

print(new_text)
```

```python
# Remove extra spaces from a string

text = "Hello. Python         is    great     !"

pattern = r"\s+"
cleaned = re.sub(pattern, " ", text)

print(cleaned)
```

```python
# Masking phone numbers in a string
text = "Call me at 9876543210 or 9123456789."

pattern = r"\d{6}(\d{4})"

masked = re.sub(pattern, r"******\1", text)

print(masked)
```

```python
# Remove HTML tags from a string
html = "<div>Hello</div><p>World</p>"

pattern = r"<.*?>"
text = re.sub(pattern, " ", html)
print(text)
```

```python
text = """
Hello, my     email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit     https://example.com for more info.
Call me   at +91-9876543210     tomorrow.
#NLP is fun! #AI #MachineLearning
"""

# Remove URLs
clean_text = re.sub(r"http\S+", "", text)

# Remove special characters
clean_text = re.sub(r"[^\w\s]", "", clean_text)

# Normalize spaces
clean_text = re.sub(r"\s+", " ", clean_text)

print(clean_text)
```

# 7️⃣ Full Match → Strict validation

```python
number = "9876543210" # 98765430 

pattern = r"\d{10}"   # exactly 10 digits

if re.fullmatch(pattern, number):
    print("Valid number")
else:
    print("Invalid number")
```

```python
pattern = r"\d+"

output = re.fullmatch(pattern, "123456") # works
# output = re.fullmatch(pattern, "123456 and 678123") # ERROR

print(output.group())
```

```python
pattern = r"\d+" # Should be None since it doesn't match the entire string

output = re.fullmatch(pattern, "123abc")

print(output)  
```

# 8️ Compile → Reusable + Efficient

```python
text = """
Hello, my email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit https://example.com for more info.
Call me at +91-9876543210 tomorrow.
#NLP is fun! #AI #MachineLearning
"""

email_pattern = re.compile(r"\b\w+@\w+\.\w+\b")

emails = email_pattern.findall(text)
print(emails)
```

# 🔥 Mini NLP Pipeline (Combine Everything)

```python
text = """
Hello, my email is john.doe@gmail.com and my backup is jd123@yahoo.com.
Visit https://example.com for more info.
Call me at +91-9876543210 tomorrow.
#NLP is fun! #AI #MachineLearning
"""

def clean_text_pipeline(text):
    text = re.sub(r"http\S+", "", text)          # remove URLs
    text = re.sub(r"[^\w\s#]", "", text)         # keep hashtags
    text = re.sub(r"\s+", " ", text)             # normalize spaces
    return text.strip()

def extract_info(text):
    emails = re.findall(r"\b\w+@\w+\.\w+\b", text)
    hashtags = re.findall(r"#\w+", text)
    numbers = re.findall(r"\d+", text)
    
    return {
        "emails": emails,
        "hashtags": hashtags,
        "numbers": numbers
    }

cleaned = clean_text_pipeline(text)
info = extract_info(text)

print("Cleaned Text:\n", cleaned)
print("\nExtracted Info:\n", info)
```

```python
# ============================================
# GENERATOR FUNCTION
# ============================================

def simple_generator():
    yield 1
    yield 2
    yield 3

gen = simple_generator()

print(next(gen))
print(next(gen))
print(next(gen))
```

# Generator with Loop

```python
def count_up_to(n):
    for i in range(n):
        yield i


for value in count_up_to(5):
    print(value)  # Function remembers its state between calls.
```

# Memory Efficient Programming

```python
# List (Consumes Memory)
# Creates ALL numbers in memory

numbers = [x*x for x in range(1000000)]

print(type(numbers))
```

```python
# Generator Expression (Memory Efficient)
# Generates values one-by-one

numbers_gen = (x*x for x in range(1000000))

print(type(numbers_gen))
```

```python
# Only computed when needed

print(next(numbers_gen))
print(next(numbers_gen))
```

```python
# Memory Comparison
import sys

list_obj = [i for i in range(10000)]
gen_obj  = (i for i in range(10000))

print("List memory     :", sys.getsizeof(list_obj))
print("Generator memory:", sys.getsizeof(gen_obj))
```

# Real-World Example — Reading Large File

```python
# 1) create a file
filename='junk.txt'

file = open(filename,'w')
file.write("Hello, loxfordacademy.com" * 5)
file.close()
```

```python
# BAD: loads entire file into memory

def read_file_bad(filename):
    with open(filename) as f:
        return f.readlines()


# Usage
for line in read_file_bad(filename):
    # process(line)
    print(line)
```

```python
# GOOD: generator reads line by line

def read_file_good(filename):
    with open(filename) as f:
        for line in f:
            yield line

# Usage
for line in read_file_good(filename):
    # process(line)
    print(line)
```

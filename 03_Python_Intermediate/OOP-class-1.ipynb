# 1. What is Object-Oriented Programming?

```python
# ======================================================
# OBJECT ORIENTED PROGRAMMING (OOP)
# ======================================================

# OOP organizes code using:
# 1. Classes  -> Blueprint
# 2. Objects  -> Real instances created from class

# Example:
# Class = Car blueprint
# Object = Actual car
```

#  2. Creating a Class

````python
# Define a simple class

class Student:
    pass   # pass means empty block (placeholder)


# Creating objects (instances)
s1 = Student()
s2 = Student()

print(s1)
print(s2)

# Each object has a different memory location
````

# 👤 3. Attributes (Variables inside Class)

````python
# Attributes store data related to an object

class Student:
    
    # constructor method
    def __init__(self, name, age):
        self.name = name   # instance attribute
        self.age = age

# Creating objects
student1 = Student("Ash", 25)
student2 = Student("Riya", 22)

print(student1.name)
print(student1.age)

print(student2.name)
print(student2.age)
````

## Important Concept

self refers to the CURRENT object

- student1.name  → belongs only to student1
- student2.name  → belongs only to student2


# 4. Methods (Functions inside Class)

````python
class Student:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # method (function inside class)
    def introduce(self):
        print("Hi, my name is", self.name)


s = Student("Rahul", 24)

# calling method using object
s.introduce()
````


## Multiple Methods Example


````python
class Calculator:
    
    def add(self, a, b):
        return a + b
    
    def multiply(self, a, b):
        return a * b

calc = Calculator()

print(calc.add(3, 5))
print(calc.multiply(4, 6))
````

````python
```

---

# 🏁 5. Constructors (`__init__`)

```python
# Constructor runs automatically when object is created

class Car:
    
    def __init__(self, brand, price):
        print("Constructor called!")
        self.brand = brand
        self.price = price
```

---

```python
c1 = Car("Toyota", 2000000)

print(c1.brand)
print(c1.price)
```

---

## Default Constructor Values

```python
class Employee:
    
    def __init__(self, name, salary=30000):
        self.name = name
        self.salary = salary
```

---

```python
e1 = Employee("Aman")
e2 = Employee("Neha", 50000)

print(e1.salary)
print(e2.salary)
```

---

# 📦 6. Instance Variables vs Class Variables

---

## ✅ Instance Variables

```python
class Dog:
    
    def __init__(self, name):
        self.name = name   # instance variable
```

---

```python
d1 = Dog("Tommy")
d2 = Dog("Rocky")

print(d1.name)
print(d2.name)

# Each object has its OWN copy
```

---

## ✅ Class Variables

```python
class Dog:
    
    species = "Canine"   # class variable (shared)
    
    def __init__(self, name):
        self.name = name
```

---

```python
d1 = Dog("Tommy")
d2 = Dog("Rocky")

print(d1.species)
print(d2.species)

# Same value shared across objects
```

---

## Accessing Class Variable

```python
print(Dog.species)   # accessed using class name
```

---

## Changing Class Variable

```python
Dog.species = "Domestic Canine"

print(d1.species)
print(d2.species)
```

---

# 🔬 Instance vs Class Variable (Important Demo)

```python
class Counter:
    
    count = 0   # class variable
    
    def __init__(self):
        Counter.count += 1   # increment shared counter
```

---

```python
c1 = Counter()
c2 = Counter()
c3 = Counter()

print("Total objects created:", Counter.count)
```

👉 Used for:

* counting objects
* shared configuration
* constants

---

# 🎯 Real Example (Student Management)

```python
class Student:
    
    school_name = "ABC School"   # class variable
    
    def __init__(self, name, marks):
        self.name = name          # instance variable
        self.marks = marks
    
    def display(self):
        print(self.name, self.marks, Student.school_name)
```

---

```python
s1 = Student("Ash", 90)
s2 = Student("Riya", 85)

s1.display()
s2.display()
```

---

# OOP Recap

```python
# CLASS      -> Blueprint
# OBJECT     -> Instance of class
# ATTRIBUTE  -> Data (variables)
# METHOD     -> Function inside class
# CONSTRUCTOR-> __init__()
# INSTANCE VARIABLE -> Unique per object
# CLASS VARIABLE    -> Shared by all objects
```

---

# Practice Exercise 

```python
# Create a class "BankAccount"
# Attributes:
#   account_holder
#   balance
#
# Methods:
#   deposit(amount)
#   withdraw(amount)
#   show_balance()
```

````

### **Exact steps** to create a **SendGrid API Key** (you only need to do this once).

---

## Step-by-Step: Get SendGrid API Key

### 1️ Create a SendGrid Account

1. Go to **[https://sendgrid.com](https://sendgrid.com)**
2. Click **Start for Free**
3. Sign up using:

   * Gmail / Microsoft account **OR**
   * Email + password

Verify your email address.

---

### 2 Complete Account Setup

After login, you’ll be asked to:

* Add your **name**
* Select use case → choose **Developer / Transactional**
* Add company name (can be anything for learning)

Click **Continue**.

---

### 3️ Open API Keys Page

Inside dashboard:

1. Left sidebar → **Settings**
2. Click **API Keys**

---

### 4️ Create New API Key

1. Click **Create API Key**
2. Give a name:

```
python-email-project
```

3. Choose permission:

**Full Access** (easy for beginners)

(or Restricted Access later for production)

---

### 5️ Generate the Key

Click:

```
Create & View
```

You will see something like:

```
SG.xxxxxxxxxxxxxxxxxxxxxxxxx
```

**IMPORTANT**

* Copy it immediately.
* You will NOT see it again.

Save it in:

* `.env` file
* password manager
* secure note

---

### 6️ Verify Sender Email (Required)

SendGrid will NOT send emails until sender is verified.

Go to:

```
Settings → Sender Authentication
```

Then:

1. Click **Verify a Single Sender**
2. Enter:

   * Name
   * Your email
   * Address (can be personal for testing)
3. Click **Create**
4. Confirm via email link.



# Now you can send emails.

````python
from sendgrid import SendGridAPIClient
from sendgrid.helpers.mail import Mail
import os


message = Mail(
    from_email='your_verified_email@gmail.com',
    to_emails='receiver@gmail.com',
    subject='Test Email',
    html_content='<strong>Hello from Python!</strong>'
)


# SENDGRID_API_KEY = os.getenv("SENDGRID_API_KEY")  # This is the industry standard practice
sg = SendGridAPIClient("SENDGRID_API_KEY") # key goes here
response = sg.send(message)

print(response.status_code)

# 202 means success
````
